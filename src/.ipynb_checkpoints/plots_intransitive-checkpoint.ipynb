```python
import os
import glob
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# =========================
# CONFIG
# =========================
BASE_DIR = r"C:\Users\mduran\Desktop\PAPERS\HOIs competitive communities\HOIs_dynamics\results"
LOGX = True
N_BINS = 15
H_TOL = 1e-12  # tolerance to compare stored H entries

PLANTS = ["Seagrasses", "Trees", "LandHerbsSaltMarsh"]


# =========================
# HELPERS: IO + CLEANING
# =========================
def _safe_read_csv(path):
    try:
        return pd.read_csv(path)
    except Exception as e:
        print(f"[WARN] Could not read {path}: {e}")
        return None


def _coerce_numeric(df, cols):
    for c in cols:
        if c in df.columns:
            df[c] = pd.to_numeric(df[c], errors="coerce")
    return df


def _ensure_pc(df):
    # p_c_i sometimes exists; if not, infer from alpha_c
    if "p_c_i" not in df.columns and "alpha_c" in df.columns:
        df["p_c_i"] = (~pd.to_numeric(df["alpha_c"], errors="coerce").isna()).astype(int)
    if "p_c_i" in df.columns:
        df["p_c_i"] = pd.to_numeric(df["p_c_i"], errors="coerce").round().clip(0, 1)
    return df


def _parse_hid_from_filename(fname):
    # expects ..._H{hid}.csv
    base = os.path.basename(fname)
    # find last occurrence of "H" followed by digits
    import re
    m = re.search(r"_H(\d+)\.csv$", base)
    if m:
        return int(m.group(1))
    return None


def load_gaussian_files(base_dir):
    # gaussian_multiH_all_H{h}.csv
    pat = os.path.join(base_dir, "gaussian_multiH_all_H*.csv")
    files = sorted(glob.glob(pat))
    if not files:
        print(f"[WARN] No gaussian files found matching: {pat}")
        return pd.DataFrame()

    frames = []
    for fp in files:
        df = _safe_read_csv(fp)
        if df is None or df.empty:
            continue
        hid = _parse_hid_from_filename(fp)
        if "H_id" not in df.columns and hid is not None:
            df["H_id"] = hid
        df["dataset"] = "gaussian"
        frames.append(df)

    out = pd.concat(frames, ignore_index=True) if frames else pd.DataFrame()
    out = _coerce_numeric(out, ["H_id", "alpha_c", "p_c_i", "std_ratios_C", "std_gaussian", "std_f", "std_d"])
    out = _ensure_pc(out)
    return out


def load_plants_files(base_dir, plant_types):
    # {plant}_realdata_multiH_all_H{h}.csv
    frames = []
    for plant in plant_types:
        pat = os.path.join(base_dir, f"{plant}_realdata_multiH_all_H*.csv")
        files = sorted(glob.glob(pat))
        if not files:
            print(f"[WARN] No plant files for {plant} matching: {pat}")
            continue

        for fp in files:
            df = _safe_read_csv(fp)
            if df is None or df.empty:
                continue
            hid = _parse_hid_from_filename(fp)
            if "H_id" not in df.columns and hid is not None:
                df["H_id"] = hid
            if "plant_type" not in df.columns:
                df["plant_type"] = plant
            df["dataset"] = "plants"
            frames.append(df)

    out = pd.concat(frames, ignore_index=True) if frames else pd.DataFrame()
    out = _coerce_numeric(out, ["H_id", "alpha_c", "p_c_i", "std_ratios_C", "std_f", "std_d"])
    out = _ensure_pc(out)
    return out


def load_meta_files(base_dir):
    g_meta_path = os.path.join(base_dir, "gaussian_multiH_Hmeta.csv")
    p_meta_path = os.path.join(base_dir, "realdata_multiH_Hmeta.csv")

    g_meta = _safe_read_csv(g_meta_path)
    p_meta = _safe_read_csv(p_meta_path)

    if g_meta is None:
        g_meta = pd.DataFrame()
    if p_meta is None:
        p_meta = pd.DataFrame()

    for df in (g_meta, p_meta):
        _coerce_numeric(df, ["H_id", "seed_H", "H01", "H02", "H12"])
        if "intransitive" in df.columns:
            # allow True/False or strings
            if df["intransitive"].dtype == object:
                df["intransitive"] = df["intransitive"].astype(str).str.lower().isin(["true", "1", "yes"])
            else:
                df["intransitive"] = df["intransitive"].astype(bool)

    return g_meta, p_meta


# =========================
# HELPERS: CONSISTENCY CHECKS
# =========================
def _compare_meta_rows(a, b, tol=1e-12):
    """
    Compare two dict-like meta rows.
    We require equal H_id and close H01/H02/H12, same seed_H if present, same intransitive if present.
    """
    keys_float = ["H01", "H02", "H12"]
    keys_int = ["H_id", "seed_H"]
    keys_bool = ["intransitive"]

    for k in keys_int:
        if k in a and k in b and pd.notna(a[k]) and pd.notna(b[k]):
            if int(a[k]) != int(b[k]):
                return False

    for k in keys_bool:
        if k in a and k in b and pd.notna(a[k]) and pd.notna(b[k]):
            if bool(a[k]) != bool(b[k]):
                return False

    for k in keys_float:
        if k in a and k in b and pd.notna(a[k]) and pd.notna(b[k]):
            if not np.isfinite(float(a[k])) or not np.isfinite(float(b[k])):
                return False
            if abs(float(a[k]) - float(b[k])) > tol:
                return False

    return True


def check_H_consistency(gauss, plants, g_meta, p_meta, tol=1e-12):
    """
    (i) Check that H entries agree between:
        - gaussian csvs (they include H01/H02/H12)
        - plants csvs (they include H01/H02/H12)
        - gaussian_meta and realdata_meta
    """
    print("\n========== CHECK (i): H consistency ==========")

    # A) gaussian H_id -> representative row
    g_rep = {}
    if not gauss.empty:
        for hid, sub in gauss.groupby("H_id"):
            # take first row
            m = sub.iloc[0].to_dict()
            g_rep[int(hid)] = m

    # B) plants H_id -> representative row (first across all plants)
    p_rep = {}
    if not plants.empty:
        for hid, sub in plants.groupby("H_id"):
            m = sub.iloc[0].to_dict()
            p_rep[int(hid)] = m

    # C) meta maps
    g_meta_map = {}
    if not g_meta.empty and "H_id" in g_meta.columns:
        for hid, sub in g_meta.groupby("H_id"):
            g_meta_map[int(hid)] = sub.iloc[0].to_dict()

    p_meta_map = {}
    if not p_meta.empty and "H_id" in p_meta.columns:
        for hid, sub in p_meta.groupby("H_id"):
            p_meta_map[int(hid)] = sub.iloc[0].to_dict()

    # Union of H ids
    all_h = sorted(set(g_rep.keys()) | set(p_rep.keys()) | set(g_meta_map.keys()) | set(p_meta_map.keys()))
    if not all_h:
        print("[WARN] No H_id found anywhere.")
        return

    mismatches = 0
    missing = 0

    for hid in all_h:
        sources = {}
        if hid in g_rep:
            sources["gaussian_csv"] = g_rep[hid]
        if hid in p_rep:
            sources["plants_csv"] = p_rep[hid]
        if hid in g_meta_map:
            sources["gaussian_meta"] = g_meta_map[hid]
        if hid in p_meta_map:
            sources["plants_meta"] = p_meta_map[hid]

        # if fewer than 2 sources, can't compare
        if len(sources) < 2:
            missing += 1
            print(f"[INFO] H{hid}: only found in {list(sources.keys())}")
            continue

        # compare everything to the first source
        keys = list(sources.keys())
        ref = sources[keys[0]]
        ok_all = True
        for k in keys[1:]:
            if not _compare_meta_rows(ref, sources[k], tol=tol):
                ok_all = False
                break

        if not ok_all:
            mismatches += 1
            print(f"[MISMATCH] H{hid}: sources differ -> {list(sources.keys())}")
            # show values to debug
            for k, row in sources.items():
                show = {kk: row.get(kk, None) for kk in ["H_id", "seed_H", "intransitive", "H01", "H02", "H12"]}
                print(f"  - {k}: {show}")

    print(f"\nSummary: total H={len(all_h)}, mismatches={mismatches}, only-one-source={missing}")
    if mismatches == 0:
        print("[OK] All comparable H entries match within tolerance.")


# =========================
# HELPERS: BINNING + SUMMARIES
# =========================
def summarize_by_bins(
    df,
    x_col="std_ratios_C",
    y_col="alpha_c",
    p_col="p_c_i",
    group_cols=("H_id",),
    n_bins=15,
    # --- new controls ---
    min_total_points_per_bin=50,      # mínimo total de puntos en el bin (por curva)
    min_Hs_per_bin=3,                 # mínimo de matrices H distintas representadas en el bin (solo si group_cols incluye H_id)
    min_points_per_H_in_bin=5,        # mínimo de puntos por H dentro del bin (solo si group_cols NO incluye H_id)
    drop_sparse_bins=True,            # si True, elimina bins que no cumplen; si False, los deja
    verbose=False
):
    """
    Summary with log-binning in x_col (x_col > 0):
      - If group_cols contains "H_id": produces per-H curves; we only enforce min_total_points_per_bin.
      - If group_cols does NOT contain "H_id" (pooled): enforces:
            * at least min_Hs_per_bin distinct H in bin
            * at least min_points_per_H_in_bin points per H in bin (optional but recommended)
            * at least min_total_points_per_bin total points in bin
    Returns columns: group_cols..., x_bin, x_center, y_mean, p_mean, n_points, n_H
    """
    d = df.copy()

    # ---- clean x ----
    if x_col not in d.columns:
        return pd.DataFrame()
    d[x_col] = pd.to_numeric(d[x_col], errors="coerce")
    d = d.dropna(subset=[x_col])
    d = d[np.isfinite(d[x_col].values)]
    d = d[d[x_col] > 0]
    if d.empty:
        return pd.DataFrame()

    # ---- y and p ----
    if y_col in d.columns:
        d[y_col] = pd.to_numeric(d[y_col], errors="coerce")
    else:
        d[y_col] = np.nan

    if p_col in d.columns:
        d[p_col] = pd.to_numeric(d[p_col], errors="coerce").round().clip(0, 1)
    else:
        d[p_col] = (~d[y_col].isna()).astype(int)

    # ---- log-bins ----
    xmin, xmax = float(d[x_col].min()), float(d[x_col].max())
    if (not np.isfinite(xmin)) or (not np.isfinite(xmax)) or xmin <= 0 or xmax <= 0 or xmin == xmax:
        d["x_bin"] = "all"
        d["x_center"] = float(d[x_col].median())
    else:
        edges = np.logspace(np.log10(xmin), np.log10(xmax), n_bins + 1)
        d["x_bin"] = pd.cut(d[x_col], bins=edges, include_lowest=True)
        d = d.dropna(subset=["x_bin"])
        if d.empty:
            return pd.DataFrame()
        intervals = d["x_bin"].cat.categories
        centers = np.sqrt(intervals.left.to_numpy() * intervals.right.to_numpy())
        center_map = {intervals[i]: float(centers[i]) for i in range(len(intervals))}
        d["x_center"] = d["x_bin"].map(center_map)

    group_cols = tuple(group_cols) if group_cols else tuple()
    gb_cols = list(group_cols) + ["x_bin"]

    # ---- first pass aggregate (means + counts) ----
    out = (
        d.groupby(gb_cols, dropna=False)
         .agg(
            x_center=("x_center", "first"),
            y_mean=(y_col, "mean"),
            p_mean=(p_col, "mean"),
            n_points=(x_col, "size"),
         )
         .reset_index()
    )

    # Always try to compute how many H contribute to each bin (even per-H, it'll be 1)
    if "H_id" in d.columns:
        # count distinct H within each (group_cols_without_H + x_bin)
        gb_for_nH = [c for c in group_cols if c != "H_id"] + ["x_bin"]
        if gb_for_nH:
            nH = d.groupby(gb_for_nH)["H_id"].nunique().reset_index().rename(columns={"H_id": "n_H"})
            out = out.merge(nH, on=gb_for_nH, how="left")
        else:
            # fully pooled: just per x_bin
            nH = d.groupby(["x_bin"])["H_id"].nunique().reset_index().rename(columns={"H_id": "n_H"})
            out = out.merge(nH, on=["x_bin"], how="left")
    else:
        out["n_H"] = np.nan

    # ---- support filters ----
    # Case A: per-H curves (group_cols includes H_id) -> just require n_points >= min_total_points_per_bin
    if "H_id" in group_cols:
        mask = out["n_points"] >= min_total_points_per_bin
        if drop_sparse_bins:
            out = out[mask].copy()
        elif verbose:
            dropped = (~mask).sum()
            print(f"[INFO] per-H: bins failing min_total_points_per_bin={min_total_points_per_bin}: {dropped}")

    # Case B: pooled curves (group_cols does NOT include H_id) -> require enough H support + enough points per H
    else:
        # i) min_total_points_per_bin
        mask_total = out["n_points"] >= min_total_points_per_bin

        # ii) min_Hs_per_bin (requires H_id in df)
        if "H_id" in d.columns and "n_H" in out.columns:
            mask_nH = out["n_H"] >= min_Hs_per_bin
        else:
            mask_nH = True

        # iii) min_points_per_H_in_bin: enforce per (x_bin, H_id) counts
        if "H_id" in d.columns:
            # count points per (x_bin, H)
            perH = (
                d.groupby(["x_bin", "H_id"])
                 .size()
                 .reset_index(name="n_per_H")
            )
            # for each bin: how many H satisfy n_per_H >= min_points_per_H_in_bin?
            per_bin_goodH = (
                perH[perH["n_per_H"] >= min_points_per_H_in_bin]
                .groupby("x_bin")["H_id"]
                .nunique()
                .reset_index(name="n_H_with_minPts")
            )
            out = out.merge(per_bin_goodH, on="x_bin", how="left")
            out["n_H_with_minPts"] = out["n_H_with_minPts"].fillna(0)

            mask_minPts = out["n_H_with_minPts"] >= min_Hs_per_bin
        else:
            mask_minPts = True

        mask = mask_total & mask_nH & mask_minPts
        if drop_sparse_bins:
            out = out[mask].copy()
        elif verbose:
            dropped = (~mask).sum()
            print(f"[INFO] pooled: bins failing support filters: {dropped}")

    # ---- sort ----
    sort_cols = list(group_cols) + ["x_center"] if group_cols else ["x_center"]
    return out.sort_values(sort_cols).reset_index(drop=True)


# =========================
# HELPERS: PLOTTING (ROBUST)
# =========================
def plot_summary_perH_generic(summary, y_col, title, ylabel, logx=True, h_col="H_id", savepath=None):
    if summary is None or summary.empty:
        print(f"[WARN] Empty summary: {title}")
        return
    if y_col not in summary.columns:
        raise KeyError(f"{y_col} not in summary columns: {list(summary.columns)}")

    plt.figure(figsize=(5.5,3.25))
    for hid, sub in summary.groupby(h_col):
        sub = sub.sort_values("x_center")
        x = pd.to_numeric(sub["x_center"], errors="coerce").to_numpy(dtype=float)
        y = pd.to_numeric(sub[y_col], errors="coerce").to_numpy(dtype=float)
        ok = np.isfinite(x) & np.isfinite(y)
        if np.any(ok):
            plt.plot(x[ok], y[ok], marker="o", linewidth=1, label=f"H {int(hid)}")

    plt.title(title)
    plt.xlabel("std_ratios_C")
    if logx:
        plt.xscale("log")
    plt.ylabel(ylabel)
    plt.legend(
    loc="center left",
    bbox_to_anchor=(1.02, 0.5),
    ncol=2,
    fontsize=6,
    frameon=False,
    )
    plt.tight_layout()
    if savepath is not None:
        save_current_fig(savepath)
    plt.show()


def plot_summary_pooled_generic(summary, y_col, title, ylabel, logx=True, savepath=None):
    if summary is None or summary.empty:
        print(f"[WARN] Empty summary: {title}")
        return
    if y_col not in summary.columns:
        raise KeyError(f"{y_col} not in summary columns: {list(summary.columns)}")

    plt.figure(figsize=(5,3.5))
    sub = summary.sort_values("x_center")
    x = pd.to_numeric(sub["x_center"], errors="coerce").to_numpy(dtype=float)
    y = pd.to_numeric(sub[y_col], errors="coerce").to_numpy(dtype=float)

    ok = np.isfinite(x) & np.isfinite(y)
    if np.any(ok):
        plt.plot(x[ok], y[ok], marker="o", linewidth=2)

    plt.title(title)
    plt.xlabel("std_ratios_C")
    if logx:
        plt.xscale("log")
    plt.ylabel(ylabel)
    plt.tight_layout()
    
    if savepath is not None:
        save_current_fig(savepath)

    plt.show()

# =========================
# SAVE FIGURES CONFIG
# =========================
SAVE_PLOTS = True
PLOTS_DIR = os.path.join(BASE_DIR, "plots")
os.makedirs(PLOTS_DIR, exist_ok=True)

def save_current_fig(path, dpi=300):
    if not SAVE_PLOTS:
        return
    plt.savefig(path, dpi=dpi, bbox_inches="tight")


# =========================
# MAIN: LOAD DATA
# =========================
os.chdir(BASE_DIR)

gauss = load_gaussian_files(BASE_DIR)
plants = load_plants_files(BASE_DIR, PLANTS)
g_meta, p_meta = load_meta_files(BASE_DIR)

print("Loaded:")
print(f"  gaussian rows: {len(gauss)}")
print(f"  plants rows:   {len(plants)}")
print(f"  g_meta rows:   {len(g_meta)}")
print(f"  p_meta rows:   {len(p_meta)}")

# =========================
# (ii) alpha_c vs std_ratios_C per H separately
# =========================
print("\n========== PLOTS (ii): alpha_c vs std_ratios_C per H ==========")

# --- GAUSSIAN per H ---
sum_g_perH = summarize_by_bins(
    gauss,
    x_col="std_ratios_C",
    y_col="alpha_c",
    p_col="p_c_i",
    group_cols=("H_id",),
    n_bins=N_BINS,
    min_total_points_per_bin=10,   # ajusta (10-30 típico)
    drop_sparse_bins=True,
)

plot_summary_perH_generic(
    sum_g_perH,
    y_col="y_mean",
    title="GAUSSIAN",
    ylabel="mean alpha_c",
    logx=LOGX,
    savepath=os.path.join(PLOTS_DIR,"gaussian_perH_alphac_vs_var_ratios_C_intransitive.svg")
)

# --- PLANTS per H (separate by plant_type) ---
for plant in PLANTS:
    d = plants[plants["plant_type"] == plant]
    s = summarize_by_bins(
        d,
        x_col="std_ratios_C",
        y_col="alpha_c",
        p_col="p_c_i",
        group_cols=("H_id",),
        n_bins=N_BINS,
        min_total_points_per_bin=10,   # 👈 añade también aquí
        drop_sparse_bins=True,
    )
    plot_summary_perH_generic(
        s,
        y_col="y_mean",
        title=f"{plant}",
        ylabel="mean alpha_c",
        logx=LOGX,
        savepath=os.path.join(PLOTS_DIR,plant+"_perH_alphac_vs_var_ratios_C_intransitive.svg")
    )

# =========================
# (iii) p_c vs std_ratios_C per H separately
# =========================
print("\n========== PLOTS (iii): p_c vs std_ratios_C per H ==========")

# --- GAUSSIAN per H ---
plot_summary_perH_generic(
    sum_g_perH,
    y_col="p_mean",
    title="GAUSSIAN",
    ylabel="mean p_c",
    logx=LOGX,
    savepath=os.path.join(PLOTS_DIR,"gaussian_perH_pcoex_vs_var_ratios_C_intransitive.svg")
)

# --- PLANTS per H ---
for plant in PLANTS:
    d = plants[plants["plant_type"] == plant]
    s = summarize_by_bins(
        d,
        x_col="std_ratios_C",
        y_col="alpha_c",
        p_col="p_c_i",
        group_cols=("H_id",),
        n_bins=N_BINS,
        min_total_points_per_bin=10,  # 
        drop_sparse_bins=True,
    )
    plot_summary_perH_generic(
        s,
        y_col="p_mean",
        title=f"{plant}",
        ylabel="mean p_c",
        logx=LOGX,
        savepath=os.path.join(PLOTS_DIR,plant+"_perH_pcoex_vs_var_ratios_C_intransitive.svg")
    )

# =========================
# (iv) pooled across H, separated by dataset
# =========================
print("\n========== PLOTS (iv): pooled across H, separated by dataset ==========")

# --- GAUSSIAN pooled ---
sum_g_pool = summarize_by_bins(
    gauss,
    x_col="std_ratios_C",
    y_col="alpha_c",
    p_col="p_c_i",
    group_cols=(),                 # pooled
    n_bins=N_BINS,
    min_total_points_per_bin=10,  # pooled necesita más soporte
    min_Hs_per_bin=5,
    min_points_per_H_in_bin=3,
    drop_sparse_bins=True,
)

plot_summary_pooled_generic(
    sum_g_pool,
    y_col="y_mean",
    title="GAUSSIAN",
    ylabel="mean alpha_c",
    logx=LOGX,
    savepath=os.path.join(PLOTS_DIR, "gaussian_pooled_alphac_vs_var_ratios_C_intransitive.svg"),
)
plot_summary_pooled_generic(
    sum_g_pool,
    y_col="p_mean",
    title="GAUSSIAN",
    ylabel="mean p_c",
    logx=LOGX,
    savepath=os.path.join(PLOTS_DIR, "gaussian_pooled_pcoex_vs_vs_var_ratios_C_intransitive.svg"),
)

# --- PLANTS pooled (one pooled curve per plant type) ---
for plant in PLANTS:
    d = plants[plants["plant_type"] == plant]
    s = summarize_by_bins(
        d,
        x_col="std_ratios_C",
        y_col="alpha_c",
        p_col="p_c_i",
        group_cols=(),                 # pooled
        n_bins=N_BINS,
        min_total_points_per_bin=10,  # 
        min_Hs_per_bin=5,
        min_points_per_H_in_bin=3,
        drop_sparse_bins=True,
    )
    plot_summary_pooled_generic(
        s,
        y_col="y_mean",
        title=f"{plant}",
        ylabel="mean alpha_c",
        logx=LOGX,
        savepath=os.path.join(PLOTS_DIR, plant+"_pooled_alphac_vs_var_ratios_C_intransitive.svg")
    )
    plot_summary_pooled_generic(
        s,
        y_col="p_mean",
        title=f"{plant}",
        ylabel="mean p_c",
        logx=LOGX,
        savepath=os.path.join(PLOTS_DIR, plant+"_pooled_pcoex_vs_var_ratios_C_intransitive.svg")
    )
```

# 2D phase space (H,std_ratios_C) on alpha_c

```python
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# =========================================================
# LOAD DATA
# =========================================================
BASE_DIR = r"C:\Users\mduran\Desktop\PAPERS\HOIs competitive communities\HOIs_dynamics\results"
os.chdir(BASE_DIR)

# Load all gaussian data (adapt if needed)
df = pd.concat(
    [pd.read_csv(f) for f in os.listdir(BASE_DIR) if f.startswith("gaussian_multiH_all")],
    ignore_index=True
)

# Load meta with H entries
meta = pd.read_csv("multiH_meta.csv")

# =========================================================
# BASIC CLEANING
# =========================================================
for c in ["alpha_c","std_ratios_C","H01","H02","H12"]:
    if c in df.columns:
        df[c] = pd.to_numeric(df[c], errors="coerce")

meta["H01"] = pd.to_numeric(meta["H01"], errors="coerce")
meta["H02"] = pd.to_numeric(meta["H02"], errors="coerce")
meta["H12"] = pd.to_numeric(meta["H12"], errors="coerce")

# =========================================================
# DEFINE INTRANSITIVITY METRIC
# =========================================================

def cycle_intensity(row):
    # recover H20 from antisymmetry
    H01 = row["H01"]
    H02 = row["H02"]
    H12 = row["H12"]
    
    H20 = 1 - H02
    
    return (H01 - 0.5) * (H12 - 0.5) * (H20 - 0.5)

meta["cycle_intensity"] = meta.apply(cycle_intensity, axis=1)

# Merge metric into main dataframe
df = df.merge(meta[["H_id","cycle_intensity"]], on="H_id", how="left")

# =========================================================
# REMOVE NaNs
# =========================================================
df = df.dropna(subset=["std_ratios_C","cycle_intensity","alpha_c"])

# =========================================================
# BINNING (2D PHASE SPACE)
# =========================================================

n_bins_x = 10
n_bins_y = 6

x = np.log10(df["std_ratios_C"]) #df["std_ratios_C"]
y = abs(df["cycle_intensity"])

x_bins = np.linspace(x.min(), x.max(), n_bins_x)
y_bins = np.linspace(y.min(), y.max(), n_bins_y)

df["x_bin"] = pd.cut(x, bins=x_bins)
df["y_bin"] = pd.cut(y, bins=y_bins)

heat = (
    df.groupby(["x_bin","y_bin"])["alpha_c"]
      .mean()
      .unstack()
)

# =========================================================
# PLOT HEATMAP
# =========================================================

plt.figure(figsize=(8,6))

plt.imshow(
    heat.T,
    origin="lower",
    aspect="auto",
    extent=[
        x_bins.min(), x_bins.max(),
        y_bins.min(), y_bins.max()
    ]
)

plt.colorbar(label="mean alpha_c")

plt.xlabel("std_ratios_C")
plt.ylabel("cycle_intensity (intransitivity metric)")
plt.title("Phase space: mean alpha_c")

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# -----------------------------
# CONFIG
# -----------------------------
MIN_CELL_N = 200          # <-- tune (e.g. 50 / 100 / 200)
N_BINS_X = 12
N_BINS_Y = 10

# If you already have df with columns:
# ["std_ratios_C", "alpha_c", "cycle_intensity"]
# then continue. Otherwise ensure you've merged cycle_intensity into df first.

# -----------------------------
# CLEAN
# -----------------------------
d = df.copy()
d["std_ratios_C"] = pd.to_numeric(d["std_ratios_C"], errors="coerce")
d["alpha_c"] = pd.to_numeric(d["alpha_c"], errors="coerce")
d["cycle_intensity"] = pd.to_numeric(d["cycle_intensity"], errors="coerce")
d = d.dropna(subset=["std_ratios_C", "alpha_c", "cycle_intensity"])

# log-x (recommended)
d = d[d["std_ratios_C"] > 0]
d["log_stdC"] = np.log10(d["std_ratios_C"])

# -----------------------------
# BINNING
# -----------------------------
x = d["log_stdC"].to_numpy()
y = d["cycle_intensity"].to_numpy()

x_edges = np.linspace(np.nanmin(x), np.nanmax(x), N_BINS_X + 1)
y_edges = np.linspace(np.nanmin(y), np.nanmax(y), N_BINS_Y + 1)

d["xbin"] = pd.cut(d["log_stdC"], bins=x_edges, include_lowest=True)
d["ybin"] = pd.cut(d["cycle_intensity"], bins=y_edges, include_lowest=True)

# -----------------------------
# AGGREGATE: mean alpha + counts
# -----------------------------
g = (
    d.groupby(["xbin", "ybin"], dropna=False)
     .agg(mean_alpha=("alpha_c", "mean"),
          n=("alpha_c", "size"))
     .reset_index()
)

mean_grid = g.pivot(index="ybin", columns="xbin", values="mean_alpha")
count_grid = g.pivot(index="ybin", columns="xbin", values="n")

# Ensure same ordering
mean_grid = mean_grid.sort_index(axis=0).sort_index(axis=1)
count_grid = count_grid.reindex(index=mean_grid.index, columns=mean_grid.columns)

# Mask low-occupancy cells
masked_mean = mean_grid.mask(count_grid < MIN_CELL_N)

# -----------------------------
# PLOT: counts + masked mean
# -----------------------------
fig, axes = plt.subplots(1, 2, figsize=(14, 5), constrained_layout=True)

# Panel 1: counts
im0 = axes[0].imshow(
    count_grid.to_numpy(),
    origin="lower",
    aspect="auto",
)
axes[0].set_title(f"Cell counts (n per bin)\nMIN_CELL_N = {MIN_CELL_N}")
fig.colorbar(im0, ax=axes[0], label="n")

# Panel 2: masked mean alpha_c
im1 = axes[1].imshow(
    masked_mean.to_numpy(),
    origin="lower",
    aspect="auto",
)
axes[1].set_title("Mean α_c (masked where n < MIN_CELL_N)")
fig.colorbar(im1, ax=axes[1], label="mean α_c")

# Axis labels: show a few ticks only (otherwise unreadable)
def set_bin_ticks(ax, x_edges, y_edges, n_xticks=6, n_yticks=6):
    nx = len(x_edges) - 1
    ny = len(y_edges) - 1
    xt = np.linspace(0, nx - 1, n_xticks).astype(int)
    yt = np.linspace(0, ny - 1, n_yticks).astype(int)

    ax.set_xticks(xt)
    ax.set_yticks(yt)

    # tick labels = bin centers
    x_centers = 0.5 * (x_edges[:-1] + x_edges[1:])
    y_centers = 0.5 * (y_edges[:-1] + y_edges[1:])

    ax.set_xticklabels([f"{x_centers[i]:.1f}" for i in xt])
    ax.set_yticklabels([f"{y_centers[i]:.3f}" for i in yt])

set_bin_ticks(axes[0], x_edges, y_edges)
set_bin_ticks(axes[1], x_edges, y_edges)

for ax in axes:
    ax.set_xlabel("log10(std_ratios_C)")
    ax.set_ylabel("cycle_intensity")

plt.show()

# -----------------------------
# PRINT DIAGNOSTICS
# -----------------------------
total_cells = count_grid.size
nonempty_cells = int((count_grid.fillna(0) > 0).sum().sum())
kept_cells = int((count_grid.fillna(0) >= MIN_CELL_N).sum().sum())

print("\n=== occupancy diagnostics ===")
print("total cells:", total_cells)
print("cells with n>0:", nonempty_cells)
print(f"cells with n>={MIN_CELL_N}:", kept_cells)
print("overall n:", len(d))
print("median cell n (among nonempty):", float(count_grid.stack().median()))
print("min cell n (among nonempty):", int(count_grid.stack().min()))
print("max cell n:", int(count_grid.stack().max()))
```

# sanity check duplicates plants species

```python
import os
import glob
import ast
import pandas as pd

BASE_DIR = r"C:\Users\mduran\Desktop\PAPERS\HOIs competitive communities\HOIs_dynamics\results"
PLANTS = ["Seagrasses", "Trees", "LandHerbsSaltMarsh"]

os.chdir(BASE_DIR)

print("\n========== CHECK DUPLICATES PER H AND PER PLANT ==========\n")

for plant in PLANTS:
    
    pattern = f"{plant}_realdata_multiH_all_H*.csv"
    files = sorted(glob.glob(pattern))
    
    if not files:
        print(f"[WARN] No files found for {plant}")
        continue

    print(f"\n---- Plant: {plant} ----")

    for fp in files:
        
        print(f"\nChecking file: {fp}")
        
        df = pd.read_csv(fp)

        if "indices" not in df.columns:
            print("  [SKIP] No 'indices' column found.")
            continue

        # Convert string representation of tuple/list to actual tuple
        def parse_indices(x):
            if isinstance(x, str):
                try:
                    return tuple(sorted(ast.literal_eval(x), reverse=True))
                except:
                    return x
            elif isinstance(x, (list, tuple)):
                return tuple(sorted(x, reverse=True))
            else:
                return x

        df["sorted_indices"] = df["indices"].apply(parse_indices)

        # Detect duplicates
        duplicates = df[df.duplicated(subset="sorted_indices", keep=False)]

        if not duplicates.empty:
            print(f"  ⚠ Found {len(duplicates)} duplicated rows.")
            print(duplicates[["indices", "sorted_indices"]].head())

            # Remove duplicates (keep first)
            cleaned_df = df.drop_duplicates(subset="sorted_indices", keep="first")

            print(f"  → After cleaning: {len(df)} → {len(cleaned_df)} rows")

            # OPTIONAL: overwrite or save cleaned version
            # cleaned_df.to_csv(fp, index=False)   # overwrite original
            # cleaned_df.to_csv(fp.replace(".csv", "_cleaned.csv"), index=False)

        else:
            print("  ✓ No duplicates found.")
```
