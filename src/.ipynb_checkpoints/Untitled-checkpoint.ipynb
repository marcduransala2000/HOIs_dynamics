```python
import numpy as np
import matplotlib.pyplot as plt

# --- your H generators (as you wrote) ---
def random_H_grilli(S=3, rng=None):
    rng = np.random.default_rng() if rng is None else rng
    H = np.full((S, S), 0.5, dtype=float)
    for i in range(S):
        for j in range(i+1, S):
            z = rng.uniform(0.0, 1.0)
            H[i, j] = z
            H[j, i] = 1.0 - z
    return H

def is_intransitive_3(H):
    wins = H > 0.5
    c1 = wins[0,1] and wins[1,2] and wins[2,0]
    c2 = wins[0,2] and wins[2,1] and wins[1,0]
    return c1 or c2

def random_intransitive_H_3(rng=None, max_tries=10000):
    rng = np.random.default_rng() if rng is None else rng
    for _ in range(max_tries):
        H = random_H_grilli(S=3, rng=rng)
        if is_intransitive_3(H):
            return H
    raise RuntimeError("Could not sample an intransitive H within max_tries.")

# --- CHECKS ---
def check_H_constraints(H, tol=1e-12):
    ok_diag = np.allclose(np.diag(H), 0.5, atol=tol, rtol=0)
    ok_pair = np.allclose(H + H.T, 1.0, atol=tol, rtol=0)
    return ok_diag, ok_pair

def cycle_score_3(H):
    # "strength" of the two possible 3-cycles (continuous version)
    return max(H[0,1]*H[1,2]*H[2,0], H[0,2]*H[2,1]*H[1,0])

# 1) Single-sample sanity
rng = np.random.default_rng(123)
H = random_H_grilli(3, rng=rng)
print("H=\n", H)
print("diag ok, pair ok =", check_H_constraints(H))
print("rank =", np.linalg.matrix_rank(H))
print("intransitive (coarse) =", is_intransitive_3(H))
print("cycle_score =", cycle_score_3(H))

# 2) Batch checks + distributions
def batch_H_diagnostics(n=2000, force_intransitive=False, seed=123):
    rng = np.random.default_rng(seed)
    H12 = []
    ranks = []
    intrans_flags = []
    scores = []
    for _ in range(n):
        H = random_intransitive_H_3(rng=rng) if force_intransitive else random_H_grilli(3, rng=rng)
        ok_diag, ok_pair = check_H_constraints(H)
        if not (ok_diag and ok_pair):
            raise AssertionError("H violates constraints!")
        H12.append(H[0,1])
        ranks.append(np.linalg.matrix_rank(H))
        intrans_flags.append(is_intransitive_3(H))
        scores.append(cycle_score_3(H))
    return np.array(H12), np.array(ranks), np.array(intrans_flags), np.array(scores)

H12, ranks, intrans_flags, scores = batch_H_diagnostics(n=5000, force_intransitive=False, seed=123)

print("\nBatch summary (no filtering):")
print("mean(H12) ~", H12.mean(), "  (should be ~0.5)")
print("min/max(H12) =", H12.min(), H12.max())
print("fraction rank<3 =", np.mean(ranks < 3))  # should be ~0
print("fraction intransitive (coarse) =", intrans_flags.mean())
print("cycle_score mean =", scores.mean())

# Plot hist of H12
plt.figure()
plt.hist(H12, bins=40)
plt.title("Distribution of H[0,1] (should be ~Uniform[0,1])")
plt.xlabel("H[0,1]")
plt.ylabel("count")
plt.show()

# Plot cycle_score
plt.figure()
plt.hist(scores, bins=40)
plt.title("Distribution of cycle_score_3(H)")
plt.xlabel("cycle_score")
plt.ylabel("count")
plt.show()

# 3) Repeat with forced intransitivity
H12_i, ranks_i, intrans_i, scores_i = batch_H_diagnostics(n=2000, force_intransitive=True, seed=123)
print("\nBatch summary (forced intransitive):")
print("fraction intransitive (coarse) =", intrans_i.mean(), " (should be 1.0)")
print("fraction rank<3 =", np.mean(ranks_i < 3))
```

```python
#DOMINANCE MATRIX
def random_H_grilli(S=3, rng=None):
    """
    Random dominance matrix H with:
    - H_ii = 0.5
    - for i!=j: H_ij + H_ji = 1
    """
    rng = np.random.default_rng() if rng is None else rng
    H = np.full((S, S), 0.5, dtype=float)
    for i in range(S):
        for j in range(i+1, S):
            z = rng.uniform(0.0, 1.0)
            H[i, j] = z
            H[j, i] = 1.0 - z
    return H
def is_intransitive_3(H):
    # assumes S=3
    wins = H > 0.5
    # cycle 0>1,1>2,2>0 OR 0>2,2>1,1>0
    c1 = wins[0,1] and wins[1,2] and wins[2,0]
    c2 = wins[0,2] and wins[2,1] and wins[1,0]
    return c1 or c2

def random_intransitive_H_3(rng=None, max_tries=1000):
    rng = np.random.default_rng() if rng is None else rng
    for _ in range(max_tries):
        H = random_H_grilli(S=3, rng=rng)
        if is_intransitive_3(H):
            return H
    raise RuntimeError("Could not sample an intransitive H within max_tries.")

#FUNCTIONS
def F(x1,x2, f1, f2, f3):
    return f1*x1+f2*x2+f3*(1-x1-x2)
def D(x1,x2, d1, d2, d3):
    return d1*x1+d2*x2+d3*(1-x1-x2)

#DIFFERENTIAL EQUATIONS - HIGHER-ORDER AND PAIRWISE
def x_111(x1,x2,alpha, f1, f2, f3, d1, d2, d3):
    return alpha*(x1/F(x1, x2, f1, f2, f3)**3*( D(x1, x2, d1, d2, d3)*( f1*2*H[0][0]*(H[0][0]+H[0][0])*f1*x1*f1*x1 + f1*2*H[0][0]*(H[0][1]+H[0][1])*f1*x1*f2*x2 + f1*2*H[0][0]*(H[0][2]+H[0][2])*f1*x1*f3*(1-x1-x2) + f1*2*H[0][1]*(H[0][0]+H[1][0])*f2*x2*f1*x1 + f1*2*H[0][1]*(H[0][1]+H[1][1])*f2*x2*f2*x2 + f1*2*H[0][1]*(H[0][2]+H[1][2])*f2*x2*f3*(1-x1-x2) + f1*2*H[0][2]*(H[0][0]+H[2][0])*f3*(1-x1-x2)*f1*x1 + f1*2*H[0][2]*(H[0][1]+H[2][1])*f3*(1-x2-x1)*f2*x2 + f1*2*H[0][2]*(H[0][2]+H[2][2])*f3*(1-x2-x1)*f3*(1-x2-x1)  ) - d1*F(x1,x2,f1, f2, f3)**3  ) ) + (1-alpha)*(x1/F(x1, x2, f1, f2, f3)**2*( D(x1, x2, d1, d2, d3)*( f1*2*H[0][0]*f1*x1 + f1*2*H[0][1]*f2*x2 + f1*2*H[0][2]*f3*(1-x1-x2)) - d1*F(x1, x2, f1, f2, f3)**2 ) )

def x_222(x1,x2,alpha, f1, f2, f3, d1, d2, d3):
    return alpha*(x2/F(x1, x2, f1, f2, f3)**3*( D(x1, x2, d1, d2, d3)*( f2*2*H[1][0]*(H[1][0]+H[0][0])*f1*x1*f1*x1 + f2*2*H[1][0]*(H[1][1]+H[0][1])*f1*x1*f2*x2 + f2*2*H[1][0]*(H[1][2]+H[0][2])*f1*x1*f3*(1-x1-x2) + f2*2*H[1][1]*(H[1][0]+H[1][0])*f2*x2*f1*x1 + f2*2*H[1][1]*(H[1][1]+H[1][1])*f2*x2*f2*x2 + f2*2*H[1][1]*(H[1][2]+H[1][2])*f2*x2*f3*(1-x1-x2) + f2*2*H[1][2]*(H[1][0]+H[2][0])*f3*(1-x1-x2)*f1*x1 + f2*2*H[1][2]*(H[1][1]+H[2][1])*f3*(1-x2-x1)*f2*x2 + f2*2*H[1][2]*(H[1][2]+H[2][2])*f3*(1-x2-x1)*f3*(1-x2-x1)  ) - d2*F(x1,x2,f1, f2, f3)**3  ) ) + (1-alpha)*(x2/F(x1, x2, f1, f2, f3)**2*( D(x1, x2, d1, d2, d3)*( f2*2*H[1][0]*f1*x1 + f2*2*H[1][1]*f2*x2 + f2*2*H[1][2]*f3*(1-x1-x2)) - d2*F(x1, x2, f1, f2, f3)**2 ) )

#RK4 SOLVER
def rk4(nsteps, h_int, dt, x10, x20, x30,alpha,f1, f2, f3, d1, d2, d3):
    writting_period = int(dt/h_int)
        
        
    h_2 = 0.5 * h_int
    h_6 = h_int / 6
       
    data = np.zeros((4, nsteps))
        
    #INITIAL CONDITION
        
    data[0, 0] = 0 #time
    data[1, 0] = x10 
    data[2, 0] = x20 
    data[3, 0] = x30 
        
        
    x1 = x10
    x2 = x20
    x3 = x30
       
    t = 0
    
    numerical_extinction = False

    f = None
    
    if save_dynamics_equal_physiological_rates == True:
        
        f = open('RK_dynamics_equal_rates_alpha_' + str(round(alpha,4)) + '.txt','w')
        f.write(str(t) + " " + str(x1) + " " + str(x2) + " " + str(x3) + "\n")
        
    if save_dynamics_different_physiological_rates == True:
        
        f = open('RK_dynamics_different_rates_alpha_' + str(round(alpha,4)) + '.txt','w')
        f.write(str(t) + " " + str(x1) + " " + str(x2) + " " + str(x3) + "\n")
        
    
    for i in range(nsteps):
            
        for j in range(writting_period):
            
            k_11 = x_111(x1,x2,alpha,f1, f2, f3, d1, d2, d3) 
            k_12 = x_222(x1,x2,alpha,f1, f2, f3, d1, d2, d3)
        
            k_21 = x_111( (x1 + h_2 * k_11), (x2 + h_2 * k_12),alpha,f1, f2, f3, d1, d2, d3)
            k_22 = x_222( (x1 + h_2 * k_11), (x2 + h_2 * k_12),alpha,f1, f2, f3, d1, d2, d3)
                
            k_31 = x_111( (x1 + h_2 * k_21), (x2 + h_2 * k_22),alpha,f1, f2, f3, d1, d2, d3)
            k_32 = x_222( (x1 + h_2 * k_21), (x2 + h_2 * k_22),alpha,f1, f2, f3, d1, d2, d3)
                
            k_41 = x_111( (x1 + h_int * k_31), (x2 + h_int * k_32),alpha,f1, f2, f3, d1, d2, d3)
            k_42 = x_222( (x1 + h_int * k_31), (x2 + h_int * k_32),alpha,f1, f2, f3, d1, d2, d3)
    
            x1 += h_6 * (k_11 + 2 * k_21 + 2 * k_31 + k_41)
            x2 += h_6 * (k_12 + 2 * k_22 + 2 * k_32 + k_42)
            x3 = 1 - x1 - x2
                
            #print(x1,x2,x3)
                    
            t += h_int
            
        if save_dynamics_equal_physiological_rates == True or save_dynamics_different_physiological_rates == True:
            
            f.write(str(t) + " " + str(x1) + " " + str(x2) + " " + str(x3) + "\n")
            
        data[0, i]= t
        data[1, i] = x1
        data[2, i] = x2
        data[3, i] = x3
        
        # BREAK LOOP IF EXPLOTES       
        if x1 < 1e-12 or x2 < 1e-12 or x3 < 1e-12:
            numerical_extinction = True
            break

            
    if f is not None:
        f.close()
        
    return data, numerical_extinction


def rk4_alpha(nsteps, h_int, dt, x10, x20, x30, ALPHAS, f1, f2, f3, d1, d2, d3):
    AC = np.nan 
    for alpha in ALPHAS:
        #print(alpha)
        
        data, numerical_extinction = rk4(nsteps, h_int, dt, x10, x20, x30, alpha, f1, f2, f3, d1, d2, d3)
        
        
        if numerical_extinction == True:
            continue
        
        if np.isnan(data).any():
            continue

        final_x1, final_x2, final_x3 = data[1, -1], data[2, -1], data[3, -1]
        if final_x1 < 1e-3 or final_x2 < 1e-3 or final_x3 < 1e-3:
            continue
        
        Af = data[1, nsteps-int(nsteps/5):-1]  # 20% of latest points
        Ai = data[1, int(nsteps/5) : int(nsteps/5) + int(nsteps/5)]  # 20% to 40% of earliest points
        
        Af = (max(Af) - min(Af)) / 2  # amplitude of oscillations
        Ai = (max(Ai) - min(Ai)) / 2
        #print(Af, Ai)
        
        amplitude_decrease = (Af + 0.005) - Ai
        std_last = np.std(data[1, nsteps - int(nsteps / 10): -1])
        
        if amplitude_decrease < 0 or std_last < 0.01:  # final amplitude is smaller than initial or last 10% points are cte --> transition occurs
            AC = alpha
            break
    
    return AC

def f(std,mean):
    return np.random.normal(mean,std)
def d(std,mean):
    return np.random.normal(mean,std)

def run_simulation_with_gaussian_sampling(nsteps, h_int, dt, ALPHAS, num_trials, std_list, mean_f_d=1.0):
    results = []
    for std in std_list:
        
        num_trials_done = 0  # Track the number of accepted trials for this std
        
        while num_trials_done < num_trials:
            
            # Generate f1, f2, f3, d1, d2, d3 from Gaussian distributions
            f1 = f(std, mean_f_d)
            f2 = f(std, mean_f_d)
            f3 = f(std, mean_f_d)
            d1 = d(std, mean_f_d)
            d2 = d(std, mean_f_d)
            d3 = d(std, mean_f_d)
            #print(f1,f2,f3,d1,d2,d3)

            # Normalize f and d by the maximum value to avoid numerical issues
            max_fd = max(f1, f2, f3, d1, d2, d3)
            f1_norm = f1 / max_fd
            f2_norm = f2 / max_fd
            f3_norm = f3 / max_fd
            d1_norm = d1 / max_fd
            d2_norm = d2 / max_fd
            d3_norm = d3 / max_fd
            
            if any(value <= 0 for value in [f1, f2, f3, d1, d2, d3]):
                # Skip this trial and continue generating new values
                continue

            # Mean rate for adjusting nsteps if necessary
            mean_rate = (f1_norm + f2_norm + f3_norm + d1_norm + d2_norm + d3_norm) / 6
            adjusted_nsteps = int(nsteps / mean_rate)

            x10, x20, x30 = 0.32, 0.21, 0.47

            alpha_c = rk4_alpha(adjusted_nsteps, h_int, dt, x10, x20, x30, ALPHAS,
                                f1_norm, f2_norm, f3_norm, d1_norm, d2_norm, d3_norm)

            p_c_i = 1 if not np.isnan(alpha_c) else 0

            # Calculate std of f and d
            std_f = np.std([f1, f2, f3])
            std_d = np.std([d1, d2, d3])

            # Calculate ratios_C and their std
            ratios_C = [f1 / d1, f2 / d2, f3 / d3]
            std_C = np.std(ratios_C)

            results.append({
                'std_gaussian': std,
                'f': [f1, f2, f3],
                'std_f': std_f,
                'd': [d1, d2, d3],
                'std_d': std_d,
                'ratios_C': ratios_C,
                'std_ratios_C': std_C,
                'alpha_c': alpha_c,
                'p_c_i': p_c_i # 1 if there is coexistence, otherwise 0
            })
            num_trials_done += 1
            
    return results

def run_simulation_with_real_data(plant_data, nsteps, h_int, dt, ALPHAS, num_trials):
    results = []
    used_combinations = set()
    while len(results) < num_trials:

        indices = tuple(sorted(random.sample(range(len(plant_data)), 3)))
        #print(indices)
        if indices in used_combinations:
            continue
        used_combinations.add(indices)
        species_data_1 = plant_data.iloc[indices[0]]
        species_data_2 = plant_data.iloc[indices[1]]
        species_data_3 = plant_data.iloc[indices[2]]
        #print(species_data_1)
        #print(species_data_2)
        #print(species_data_3)
        max_fd = max(species_data_1[2], species_data_2[2], species_data_3[2],species_data_1[3], species_data_2[3], species_data_3[3])
        F1, D1 = species_data_1[2], species_data_1[3]
        F2, D2 = species_data_2[2], species_data_2[3]
        F3, D3 = species_data_3[2], species_data_3[3]
        f1, d1 = species_data_1[2] / max_fd, species_data_1[3] / max_fd 
        f2, d2 = species_data_2[2] / max_fd, species_data_2[3] / max_fd
        f3, d3 = species_data_3[2] / max_fd, species_data_3[3] / max_fd
        #print(f1, d1, f2, d2, f3, d3)
        
        mean_rate = (f1 + f2 + f3 + d1 + d2 + d3) / 6
        #print('mean rate', mean_rate)
        
        adjusted_nsteps = int(nsteps / mean_rate)
        #print('adjusted nsteps', adjusted_nsteps)
        
        alpha_c = rk4_alpha(adjusted_nsteps, h_int, dt, x10, x20, x30, ALPHAS, f1, f2, f3, d1, d2, d3)
        #print(alpha_c)
        p_c_i = 1 if not np.isnan(alpha_c) else 0
        
        # Calculate std of f and d separately
        std_f = np.std([F1, F2, F3])
        std_d = np.std([D1, D2, D3])
        
        # Calculate std and proportion of each pair (f_i, d_i)
        ratios_C = [F1/D1, F2/D2, F3/D3]
        std_C = np.std([F1 / D1, F2 / D2, F3 / D3])
        
        results.append({
            'indices': indices,
            'f': [F1, F2, F3],
            'std_f': std_f,
            'd': [D1, D2, D3],
            'std_d': std_d,
            'ratios_C': ratios_C,           
            'std_ratios_C': std_C,
            'alpha_c': alpha_c,
            'p_c_i': p_c_i # 1 if there is coexistence, otherwise 0
        })
        
    return results
```

```python
import numpy as np
import pandas as pd

# --- smoke-test config ---

save_dynamics_different_physiological_rates = False
save_dynamics_equal_physiological_rates = False
nsteps = 5000
h_int = 1
dt = 1
ALPHAS = np.linspace(0, 1.0, 21)  # 0.05 step
num_trials_per_std = 50
std_list = [1e-2, 1e-1, 0.5]
mean_f_d = 1.0

rng = np.random.default_rng(123)
num_H = 5

all_rows = []

for h_id in range(num_H):
    # choose H (either unrestricted or forced intransitive)
    # H = random_H_grilli(S=3, rng=rng)
    H = random_intransitive_H_3(rng=rng)

    cs = cycle_score_3(np.array(H))

    results = run_simulation_with_gaussian_sampling(
        nsteps=nsteps,
        h_int=h_int,
        dt=dt,
        ALPHAS=ALPHAS,
        num_trials=num_trials_per_std,
        std_list=std_list,
        mean_f_d=mean_f_d
    )

    # add metadata per row
    for r in results:
        r2 = dict(r)
        r2["h_id"] = h_id
        r2["cycle_score"] = cs
        all_rows.append(r2)

df_smoke = pd.DataFrame(all_rows)
df_smoke.to_csv("smoke_fig3_multiH.csv", index=False)

print(df_smoke.head())
print("\nRows:", len(df_smoke), " Unique H:", df_smoke["h_id"].nunique())
print("Fraction coexist:", df_smoke["p_c_i"].mean())
print("alpha_c (non-NaN) mean:", df_smoke.loc[df_smoke["alpha_c"].notna(), "alpha_c"].mean())
```

```python
import numpy as np
import pandas as pd
import os
import random


#DOMINANCE MATRIX
def random_H_grilli(S=3, rng=None):
    """
    Random dominance matrix H with:
    - H_ii = 0.5
    - for i!=j: H_ij + H_ji = 1
    """
    rng = np.random.default_rng() if rng is None else rng
    H = np.full((S, S), 0.5, dtype=float)
    for i in range(S):
        for j in range(i+1, S):
            z = rng.uniform(0.0, 1.0)
            H[i, j] = z
            H[j, i] = 1.0 - z
    return H
def is_intransitive_3(H):
    # assumes S=3
    wins = H > 0.5
    # cycle 0>1,1>2,2>0 OR 0>2,2>1,1>0
    c1 = wins[0,1] and wins[1,2] and wins[2,0]
    c2 = wins[0,2] and wins[2,1] and wins[1,0]
    return c1 or c2

def random_intransitive_H_3(rng=None, max_tries=1000):
    rng = np.random.default_rng() if rng is None else rng
    for _ in range(max_tries):
        H = random_H_grilli(S=3, rng=rng)
        if is_intransitive_3(H):
            return H
    raise RuntimeError("Could not sample an intransitive H within max_tries.")

def cycle_score_3(H):
    # "strength" of the two possible 3-cycles (continuous version)
    return max(H[0,1]*H[1,2]*H[2,0], H[0,2]*H[2,1]*H[1,0])

#FUNCTIONS
def F(x1,x2, f1, f2, f3):
    return f1*x1+f2*x2+f3*(1-x1-x2)
def D(x1,x2, d1, d2, d3):
    return d1*x1+d2*x2+d3*(1-x1-x2)

#DIFFERENTIAL EQUATIONS - HIGHER-ORDER AND PAIRWISE
def x_111(x1,x2,alpha, f1, f2, f3, d1, d2, d3):
    return alpha*(x1/F(x1, x2, f1, f2, f3)**3*( D(x1, x2, d1, d2, d3)*( f1*2*H[0][0]*(H[0][0]+H[0][0])*f1*x1*f1*x1 + f1*2*H[0][0]*(H[0][1]+H[0][1])*f1*x1*f2*x2 + f1*2*H[0][0]*(H[0][2]+H[0][2])*f1*x1*f3*(1-x1-x2) + f1*2*H[0][1]*(H[0][0]+H[1][0])*f2*x2*f1*x1 + f1*2*H[0][1]*(H[0][1]+H[1][1])*f2*x2*f2*x2 + f1*2*H[0][1]*(H[0][2]+H[1][2])*f2*x2*f3*(1-x1-x2) + f1*2*H[0][2]*(H[0][0]+H[2][0])*f3*(1-x1-x2)*f1*x1 + f1*2*H[0][2]*(H[0][1]+H[2][1])*f3*(1-x2-x1)*f2*x2 + f1*2*H[0][2]*(H[0][2]+H[2][2])*f3*(1-x2-x1)*f3*(1-x2-x1)  ) - d1*F(x1,x2,f1, f2, f3)**3  ) ) + (1-alpha)*(x1/F(x1, x2, f1, f2, f3)**2*( D(x1, x2, d1, d2, d3)*( f1*2*H[0][0]*f1*x1 + f1*2*H[0][1]*f2*x2 + f1*2*H[0][2]*f3*(1-x1-x2)) - d1*F(x1, x2, f1, f2, f3)**2 ) )

def x_222(x1,x2,alpha, f1, f2, f3, d1, d2, d3):
    return alpha*(x2/F(x1, x2, f1, f2, f3)**3*( D(x1, x2, d1, d2, d3)*( f2*2*H[1][0]*(H[1][0]+H[0][0])*f1*x1*f1*x1 + f2*2*H[1][0]*(H[1][1]+H[0][1])*f1*x1*f2*x2 + f2*2*H[1][0]*(H[1][2]+H[0][2])*f1*x1*f3*(1-x1-x2) + f2*2*H[1][1]*(H[1][0]+H[1][0])*f2*x2*f1*x1 + f2*2*H[1][1]*(H[1][1]+H[1][1])*f2*x2*f2*x2 + f2*2*H[1][1]*(H[1][2]+H[1][2])*f2*x2*f3*(1-x1-x2) + f2*2*H[1][2]*(H[1][0]+H[2][0])*f3*(1-x1-x2)*f1*x1 + f2*2*H[1][2]*(H[1][1]+H[2][1])*f3*(1-x2-x1)*f2*x2 + f2*2*H[1][2]*(H[1][2]+H[2][2])*f3*(1-x2-x1)*f3*(1-x2-x1)  ) - d2*F(x1,x2,f1, f2, f3)**3  ) ) + (1-alpha)*(x2/F(x1, x2, f1, f2, f3)**2*( D(x1, x2, d1, d2, d3)*( f2*2*H[1][0]*f1*x1 + f2*2*H[1][1]*f2*x2 + f2*2*H[1][2]*f3*(1-x1-x2)) - d2*F(x1, x2, f1, f2, f3)**2 ) )

#RK4 SOLVER
def rk4(nsteps, h_int, dt, x10, x20, x30,alpha,f1, f2, f3, d1, d2, d3):
    writting_period = int(dt/h_int)
        
        
    h_2 = 0.5 * h_int
    h_6 = h_int / 6
       
    data = np.zeros((4, nsteps))
        
    #INITIAL CONDITION
        
    data[0, 0] = 0 #time
    data[1, 0] = x10 
    data[2, 0] = x20 
    data[3, 0] = x30 
        
        
    x1 = x10
    x2 = x20
    x3 = x30
       
    t = 0
    
    numerical_extinction = False
    
    f = None
    
    if save_dynamics_equal_physiological_rates == True:
        
        f = open('RK_dynamics_equal_rates_alpha_' + str(round(alpha,4)) + '.txt','w')
        f.write(str(t) + " " + str(x1) + " " + str(x2) + " " + str(x3) + "\n")
        
    if save_dynamics_different_physiological_rates == True:
        
        f = open('RK_dynamics_different_rates_alpha_' + str(round(alpha,4)) + '.txt','w')
        f.write(str(t) + " " + str(x1) + " " + str(x2) + " " + str(x3) + "\n")
        
    
    for i in range(nsteps):
            
        for j in range(writting_period):
            
            k_11 = x_111(x1,x2,alpha,f1, f2, f3, d1, d2, d3) 
            k_12 = x_222(x1,x2,alpha,f1, f2, f3, d1, d2, d3)
        
            k_21 = x_111( (x1 + h_2 * k_11), (x2 + h_2 * k_12),alpha,f1, f2, f3, d1, d2, d3)
            k_22 = x_222( (x1 + h_2 * k_11), (x2 + h_2 * k_12),alpha,f1, f2, f3, d1, d2, d3)
                
            k_31 = x_111( (x1 + h_2 * k_21), (x2 + h_2 * k_22),alpha,f1, f2, f3, d1, d2, d3)
            k_32 = x_222( (x1 + h_2 * k_21), (x2 + h_2 * k_22),alpha,f1, f2, f3, d1, d2, d3)
                
            k_41 = x_111( (x1 + h_int * k_31), (x2 + h_int * k_32),alpha,f1, f2, f3, d1, d2, d3)
            k_42 = x_222( (x1 + h_int * k_31), (x2 + h_int * k_32),alpha,f1, f2, f3, d1, d2, d3)
    
            x1 += h_6 * (k_11 + 2 * k_21 + 2 * k_31 + k_41)
            x2 += h_6 * (k_12 + 2 * k_22 + 2 * k_32 + k_42)
            x3 = 1 - x1 - x2
                
            #print(x1,x2,x3)
                    
            t += h_int
            
        if (save_dynamics_equal_physiological_rates or save_dynamics_different_physiological_rates) and (f is not None):
            f.write(str(t) + " " + str(x1) + " " + str(x2) + " " + str(x3) + "\n")
            
        data[0, i]= t
        data[1, i] = x1
        data[2, i] = x2
        data[3, i] = x3
        
        # BREAK LOOP IF EXPLOTES       
        if x1 < 1e-12 or x2 < 1e-12 or x3 < 1e-12:
            numerical_extinction = True
            break

            
    if f is not None:
        f.close()
  
    return data, numerical_extinction


def rk4_alpha(nsteps, h_int, dt, x10, x20, x30, ALPHAS, f1, f2, f3, d1, d2, d3):
    AC = np.nan 
    for alpha in ALPHAS:
        #print(alpha)
        
        data, numerical_extinction = rk4(nsteps, h_int, dt, x10, x20, x30, alpha, f1, f2, f3, d1, d2, d3)
        
        
        if numerical_extinction == True:
            continue
        
        if np.isnan(data).any():
            continue

        final_x1, final_x2, final_x3 = data[1, -1], data[2, -1], data[3, -1]
        if final_x1 < 1e-3 or final_x2 < 1e-3 or final_x3 < 1e-3:
            continue
        
        Af = data[1, nsteps-int(nsteps/5):-1]  # 20% of latest points
        Ai = data[1, int(nsteps/5) : int(nsteps/5) + int(nsteps/5)]  # 20% to 40% of earliest points
        
        Af = (max(Af) - min(Af)) / 2  # amplitude of oscillations
        Ai = (max(Ai) - min(Ai)) / 2
        #print(Af, Ai)
        
        amplitude_decrease = (Af + 0.005) - Ai
        std_last = np.std(data[1, nsteps - int(nsteps / 10): -1])
        
        if amplitude_decrease < 0 or std_last < 0.01:  # final amplitude is smaller than initial or last 10% points are cte --> transition occurs
            AC = alpha
            break
    
    return AC


def f(std,mean):
    return np.random.normal(mean,std)
def d(std,mean):
    return np.random.normal(mean,std)

def run_simulation_with_gaussian_sampling(nsteps, h_int, dt, ALPHAS, num_trials, std_list, mean_f_d=1.0):
    results = []
    for std in std_list:
        
        num_trials_done = 0  # Track the number of accepted trials for this std
        
        while num_trials_done < num_trials:
            
            # Generate f1, f2, f3, d1, d2, d3 from Gaussian distributions
            f1 = f(std, mean_f_d)
            f2 = f(std, mean_f_d)
            f3 = f(std, mean_f_d)
            d1 = d(std, mean_f_d)
            d2 = d(std, mean_f_d)
            d3 = d(std, mean_f_d)
            #print(f1,f2,f3,d1,d2,d3)

            # Normalize f and d by the maximum value to avoid numerical issues
            max_fd = max(f1, f2, f3, d1, d2, d3)
            f1_norm = f1 / max_fd
            f2_norm = f2 / max_fd
            f3_norm = f3 / max_fd
            d1_norm = d1 / max_fd
            d2_norm = d2 / max_fd
            d3_norm = d3 / max_fd
            
            if any(value <= 0 for value in [f1, f2, f3, d1, d2, d3]):
                # Skip this trial and continue generating new values
                continue

            # Mean rate for adjusting nsteps if necessary
            mean_rate = (f1_norm + f2_norm + f3_norm + d1_norm + d2_norm + d3_norm) / 6
            adjusted_nsteps = int(nsteps / mean_rate)

            x10, x20, x30 = 0.32, 0.21, 0.47

            alpha_c = rk4_alpha(adjusted_nsteps, h_int, dt, x10, x20, x30, ALPHAS,
                                f1_norm, f2_norm, f3_norm, d1_norm, d2_norm, d3_norm)

            p_c_i = 1 if not np.isnan(alpha_c) else 0

            # Calculate std of f and d
            std_f = np.std([f1, f2, f3])
            std_d = np.std([d1, d2, d3])

            # Calculate ratios_C and their std
            ratios_C = [f1 / d1, f2 / d2, f3 / d3]
            std_C = np.std(ratios_C)

            results.append({
                'std_gaussian': std,
                'f': [f1, f2, f3],
                'std_f': std_f,
                'd': [d1, d2, d3],
                'std_d': std_d,
                'ratios_C': ratios_C,
                'std_ratios_C': std_C,
                'alpha_c': alpha_c,
                'p_c_i': p_c_i # 1 if there is coexistence, otherwise 0
            })
            num_trials_done += 1
            
    return results

def run_simulation_with_real_data(plant_data, nsteps, h_int, dt, ALPHAS, num_trials, x10, x20, x30):
    results = []
    used_combinations = set()
    while len(results) < num_trials:

        indices = tuple(sorted(random.sample(range(len(plant_data)), 3)))
        #print(indices)
        if indices in used_combinations:
            continue
        used_combinations.add(indices)
        species_data_1 = plant_data.iloc[indices[0]]
        species_data_2 = plant_data.iloc[indices[1]]
        species_data_3 = plant_data.iloc[indices[2]]
        #print(species_data_1)
        #print(species_data_2)
        #print(species_data_3)
        max_fd = max(species_data_1[2], species_data_2[2], species_data_3[2],species_data_1[3], species_data_2[3], species_data_3[3])
        F1, D1 = species_data_1[2], species_data_1[3]
        F2, D2 = species_data_2[2], species_data_2[3]
        F3, D3 = species_data_3[2], species_data_3[3]
        f1, d1 = species_data_1[2] / max_fd, species_data_1[3] / max_fd 
        f2, d2 = species_data_2[2] / max_fd, species_data_2[3] / max_fd
        f3, d3 = species_data_3[2] / max_fd, species_data_3[3] / max_fd
        #print(f1, d1, f2, d2, f3, d3)
        
        mean_rate = (f1 + f2 + f3 + d1 + d2 + d3) / 6
        #print('mean rate', mean_rate)
        
        adjusted_nsteps = int(nsteps / mean_rate)
        #print('adjusted nsteps', adjusted_nsteps)
        
        alpha_c = rk4_alpha(adjusted_nsteps, h_int, dt, x10, x20, x30, ALPHAS, f1, f2, f3, d1, d2, d3)
        #print(alpha_c)
        p_c_i = 1 if not np.isnan(alpha_c) else 0
        
        # Calculate std of f and d separately
        std_f = np.std([F1, F2, F3])
        std_d = np.std([D1, D2, D3])
        
        # Calculate std and proportion of each pair (f_i, d_i)
        ratios_C = [F1/D1, F2/D2, F3/D3]
        std_C = np.std([F1 / D1, F2 / D2, F3 / D3])
        
        results.append({
            'indices': indices,
            'f': [F1, F2, F3],
            'std_f': std_f,
            'd': [D1, D2, D3],
            'std_d': std_d,
            'ratios_C': ratios_C,           
            'std_ratios_C': std_C,
            'alpha_c': alpha_c,
            'p_c_i': p_c_i # 1 if there is coexistence, otherwise 0
        })
        
    return results

# ------------------------------------------------------------------------------------------------

# OPTIONS
# 1- RUN ONLY THE RK4 DYNAMICS for a specific fraction of HOIs (alpha) - EQUAL PHYSIOLOGICAL RATES 
# 2- RUN ONLY THE RK4 DYNAMICS for a specific fraction of HOIs (alpha) - DIFFERENT PHYSIOLOGICAL RATES
# 3- RUN THE MAIN CODE FOR GAUSSIAN DATA
# 4- RUN THE MAIN CODE FOR REAL DATA


os.chdir(r'C:\Users\mduran\Desktop\PAPERS\HOIs competitive communities\HOIs_dynamics')
```

```python
def rk4_alpha_debug(nsteps, h_int, dt, x10, x20, x30, ALPHAS,
                    f1, f2, f3, d1, d2, d3):
    AC = np.nan
    n_alpha_tested = 0

    for alpha in ALPHAS:
        n_alpha_tested += 1

        data, numerical_extinction = rk4(
            nsteps, h_int, dt, x10, x20, x30,
            alpha, f1, f2, f3, d1, d2, d3
        )

        if numerical_extinction:
            continue
        if np.isnan(data).any():
            continue

        final_x1, final_x2, final_x3 = data[1, -1], data[2, -1], data[3, -1]
        if final_x1 < 1e-3 or final_x2 < 1e-3 or final_x3 < 1e-3:
            continue

        Af = data[1, nsteps-int(nsteps/5):-1]
        Ai = data[1, int(nsteps/5): int(nsteps/5) + int(nsteps/5)]

        Af = (max(Af) - min(Af)) / 2
        Ai = (max(Ai) - min(Ai)) / 2

        amplitude_decrease = (Af + 0.005) - Ai
        std_last = np.std(data[1, nsteps - int(nsteps/10): -1])

        if amplitude_decrease < 0 or std_last < 0.01:
            AC = alpha
            break

    return AC, n_alpha_tested
import time

def run_simulation_with_gaussian_sampling_debug(nsteps, h_int, dt, ALPHAS,
                                                num_trials, std_list, mean_f_d=1.0):
    results = []
    for std in std_list:
        num_trials_done = 0
        while num_trials_done < num_trials:
            f1 = np.random.normal(mean_f_d, std)
            f2 = np.random.normal(mean_f_d, std)
            f3 = np.random.normal(mean_f_d, std)
            d1 = np.random.normal(mean_f_d, std)
            d2 = np.random.normal(mean_f_d, std)
            d3 = np.random.normal(mean_f_d, std)

            if any(v <= 0 for v in [f1,f2,f3,d1,d2,d3]):
                continue

            max_fd = max(f1, f2, f3, d1, d2, d3)
            f1n,f2n,f3n = f1/max_fd, f2/max_fd, f3/max_fd
            d1n,d2n,d3n = d1/max_fd, d2/max_fd, d3/max_fd

            mean_rate = (f1n+f2n+f3n+d1n+d2n+d3n)/6
            adjusted_nsteps = int(nsteps / mean_rate)

            x10, x20, x30 = 0.32, 0.21, 0.47

            t0 = time.time()
            alpha_c, n_alpha_tested = rk4_alpha_debug(
                adjusted_nsteps, h_int, dt, x10, x20, x30, ALPHAS,
                f1n,f2n,f3n,d1n,d2n,d3n
            )
            t1 = time.time()

            results.append({
                "std_gaussian": std,
                "alpha_c": alpha_c,
                "p_c_i": 0 if np.isnan(alpha_c) else 1,
                "adjusted_nsteps": adjusted_nsteps,
                "n_alpha_tested": n_alpha_tested,
                "sec_this_row": (t1 - t0),
            })
            num_trials_done += 1
    return results
```

```python
import numpy as np
import pandas as pd
import time

# Parámetros de test
nsteps = 5000
h_int = 1
dt = 1
ALPHAS = np.linspace(0, 1.0, 81)
std_list = [0.0001, 0.01, 0.1]   # acorta para test
num_trials_per_std = 30          # acorta para test

def summarize(df, label):
    print("\n===", label, "===")
    print("rows:", len(df))
    print("mean sec/row:", df["sec_this_row"].mean())
    print("median sec/row:", df["sec_this_row"].median())
    print("mean n_alpha_tested:", df["n_alpha_tested"].mean())
    print("median n_alpha_tested:", df["n_alpha_tested"].median())
    print("fraction alpha_c NaN:", df["alpha_c"].isna().mean())
    print("alpha_c mean (non-NaN):", df.loc[df["alpha_c"].notna(), "alpha_c"].mean())

# --- Caso 1: H fija (la antigua) ---
H = np.array([[0.5, 0.34, 0.76],
              [0.66, 0.5, 0.25],
              [0.24, 0.75, 0.5]])

res_fixed = run_simulation_with_gaussian_sampling_debug(
    nsteps, h_int, dt, ALPHAS,
    num_trials_per_std, std_list, mean_f_d=1.0
)
df_fixed = pd.DataFrame(res_fixed)
summarize(df_fixed, "Fixed H (old)")

# --- Caso 2: H aleatoria intransitiva ---
rng = np.random.default_rng(123)
H = random_intransitive_H_3(rng=rng, max_tries=10000)
H = np.array(H)

res_rand = run_simulation_with_gaussian_sampling_debug(
    nsteps, h_int, dt, ALPHAS,
    num_trials_per_std, std_list, mean_f_d=1.0
)
df_rand = pd.DataFrame(res_rand)
summarize(df_rand, "Random intransitive H")
```

```python
def rk4_alpha_trace(nsteps, h_int, dt, x10, x20, x30, ALPHAS,
                    f1,f2,f3,d1,d2,d3, max_store=10):

    trace = []
    AC = np.nan

    for alpha in ALPHAS:
        data, numerical_extinction = rk4(nsteps,h_int,dt,x10,x20,x30,alpha,f1,f2,f3,d1,d2,d3)

        if numerical_extinction or np.isnan(data).any():
            trace.append((alpha, "fail", np.nan, np.nan, np.nan))
            continue

        final_x1, final_x2, final_x3 = data[1,-1], data[2,-1], data[3,-1]
        if final_x1 < 1e-3 or final_x2 < 1e-3 or final_x3 < 1e-3:
            trace.append((alpha, "ext", np.nan, np.nan, np.nan))
            continue

        Af = data[1, nsteps-int(nsteps/5):-1]
        Ai = data[1, int(nsteps/5): int(nsteps/5) + int(nsteps/5)]
        Af = (max(Af)-min(Af))/2
        Ai = (max(Ai)-min(Ai))/2
        amplitude_decrease = (Af + 0.005) - Ai
        std_last = np.std(data[1, nsteps-int(nsteps/10):-1])

        ok = (amplitude_decrease < 0) or (std_last < 0.01)
        trace.append((alpha, "ok" if ok else "no", Af, Ai, std_last))

        if ok:
            AC = alpha
            break

        if len(trace) >= max_store and alpha > 0.2:
            # opcional: no guardes infinito
            pass

    return AC, trace
```

```python
H = np.array([[0.5, 0.34, 0.76],
              [0.66, 0.5, 0.25],
              [0.24, 0.75, 0.5]])

# fija un set de rates conflictivo (por ejemplo std grande)
f1,f2,f3,d1,d2,d3 = 1.0, 0.9, 1.04, 1.2, 0.97, 0.9
print('f1,f2,f3,d1,d2,d3=',f1,f2,f3,d1,d2,d3)
max_fd = max(f1,f2,f3,d1,d2,d3)
f1,f2,f3,d1,d2,d3 = f1/max_fd, f2/max_fd, f3/max_fd, d1/max_fd, d2/max_fd, d3/max_fd

x10,x20,x30 = 0.32,0.21,0.47
AC, trace = rk4_alpha_trace(5000, 1, 1, x10,x20,x30, np.linspace(0,1,81),
                            f1,f2,f3,d1,d2,d3)

print("alpha_c:", AC)
for row in trace[:15]:
    print(row)
```
