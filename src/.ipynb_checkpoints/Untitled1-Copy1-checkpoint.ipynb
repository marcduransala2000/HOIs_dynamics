```python
import numpy as np
import pandas as pd
import os
import random


#DOMINANCE MATRIX
def random_H_grilli(S=3, rng=None):
    """
    Grilli-style random dominance matrix H:
      H_ii = 0.5
      for i!=j: H_ij + H_ji = 1
      with H_ij ~ Uniform(0,1) for i<j
    """
    rng = np.random.default_rng() if rng is None else rng
    H = np.full((S, S), 0.5, dtype=float)
    for i in range(S):
        for j in range(i+1, S):
            z = rng.uniform(0.0, 1.0)
            H[i, j] = z
            H[j, i] = 1.0 - z
    return H

def is_intransitive_3(H, thr=0.5):
    """
    For S=3: returns True iff the coarse-grained dominance graph (H_ij > thr)
    contains a 3-cycle
    """
    wins = H > thr
    c1 = wins[0,1] and wins[1,2] and wins[2,0]
    c2 = wins[0,2] and wins[2,1] and wins[1,0]
    return bool(c1 or c2)

def random_intransitive_H_3(rng=None, max_tries=1000):
    rng = np.random.default_rng() if rng is None else rng
    for _ in range(max_tries):
        H = random_H_grilli(S=3, rng=rng)
        if is_intransitive_3(H):
            return H
    raise RuntimeError("Could not sample an intransitive H within max_tries.")

def transitivity_label_3(H, thr=0.5):
    return "intransitive" if is_intransitive_3(H, thr=thr) else "transitive"

#FUNCTIONS
def F(x1,x2, f1, f2, f3):
    return f1*x1+f2*x2+f3*(1-x1-x2)
def D(x1,x2, d1, d2, d3):
    return d1*x1+d2*x2+d3*(1-x1-x2)

#DIFFERENTIAL EQUATIONS - HIGHER-ORDER AND PAIRWISE
def x_111(x1,x2,alpha, f1, f2, f3, d1, d2, d3, H):
    return alpha*(x1/F(x1, x2, f1, f2, f3)**3*( D(x1, x2, d1, d2, d3)*( f1*2*H[0][0]*(H[0][0]+H[0][0])*f1*x1*f1*x1 + f1*2*H[0][0]*(H[0][1]+H[0][1])*f1*x1*f2*x2 + f1*2*H[0][0]*(H[0][2]+H[0][2])*f1*x1*f3*(1-x1-x2) + f1*2*H[0][1]*(H[0][0]+H[1][0])*f2*x2*f1*x1 + f1*2*H[0][1]*(H[0][1]+H[1][1])*f2*x2*f2*x2 + f1*2*H[0][1]*(H[0][2]+H[1][2])*f2*x2*f3*(1-x1-x2) + f1*2*H[0][2]*(H[0][0]+H[2][0])*f3*(1-x1-x2)*f1*x1 + f1*2*H[0][2]*(H[0][1]+H[2][1])*f3*(1-x2-x1)*f2*x2 + f1*2*H[0][2]*(H[0][2]+H[2][2])*f3*(1-x2-x1)*f3*(1-x2-x1)  ) - d1*F(x1,x2,f1, f2, f3)**3  ) ) + (1-alpha)*(x1/F(x1, x2, f1, f2, f3)**2*( D(x1, x2, d1, d2, d3)*( f1*2*H[0][0]*f1*x1 + f1*2*H[0][1]*f2*x2 + f1*2*H[0][2]*f3*(1-x1-x2)) - d1*F(x1, x2, f1, f2, f3)**2 ) )

def x_222(x1,x2,alpha, f1, f2, f3, d1, d2, d3, H):
    return alpha*(x2/F(x1, x2, f1, f2, f3)**3*( D(x1, x2, d1, d2, d3)*( f2*2*H[1][0]*(H[1][0]+H[0][0])*f1*x1*f1*x1 + f2*2*H[1][0]*(H[1][1]+H[0][1])*f1*x1*f2*x2 + f2*2*H[1][0]*(H[1][2]+H[0][2])*f1*x1*f3*(1-x1-x2) + f2*2*H[1][1]*(H[1][0]+H[1][0])*f2*x2*f1*x1 + f2*2*H[1][1]*(H[1][1]+H[1][1])*f2*x2*f2*x2 + f2*2*H[1][1]*(H[1][2]+H[1][2])*f2*x2*f3*(1-x1-x2) + f2*2*H[1][2]*(H[1][0]+H[2][0])*f3*(1-x1-x2)*f1*x1 + f2*2*H[1][2]*(H[1][1]+H[2][1])*f3*(1-x2-x1)*f2*x2 + f2*2*H[1][2]*(H[1][2]+H[2][2])*f3*(1-x2-x1)*f3*(1-x2-x1)  ) - d2*F(x1,x2,f1, f2, f3)**3  ) ) + (1-alpha)*(x2/F(x1, x2, f1, f2, f3)**2*( D(x1, x2, d1, d2, d3)*( f2*2*H[1][0]*f1*x1 + f2*2*H[1][1]*f2*x2 + f2*2*H[1][2]*f3*(1-x1-x2)) - d2*F(x1, x2, f1, f2, f3)**2 ) )

#RK4 SOLVER
def rk4(nsteps, h_int, dt, x10, x20, x30,alpha,f1, f2, f3, d1, d2, d3, H):
    writting_period = int(dt/h_int)
        
        
    h_2 = 0.5 * h_int
    h_6 = h_int / 6
       
    data = np.zeros((4, nsteps))
        
    #INITIAL CONDITION
        
    data[0, 0] = 0 #time
    data[1, 0] = x10 
    data[2, 0] = x20 
    data[3, 0] = x30 
        
        
    x1 = x10
    x2 = x20
    x3 = x30
       
    t = 0
    
    numerical_extinction = False
    
    f = None
    
    if save_dynamics_equal_physiological_rates == True:
        
        f = open('RK_dynamics_equal_rates_alpha_' + str(round(alpha,4)) + '.txt','w')
        f.write(str(t) + " " + str(x1) + " " + str(x2) + " " + str(x3) + "\n")
        
    if save_dynamics_different_physiological_rates == True:
        
        f = open('RK_dynamics_different_rates_alpha_' + str(round(alpha,4)) + '.txt','w')
        f.write(str(t) + " " + str(x1) + " " + str(x2) + " " + str(x3) + "\n")
        
    
    for i in range(nsteps):
            
        for j in range(writting_period):
            
            k_11 = x_111(x1,x2,alpha,f1, f2, f3, d1, d2, d3, H) 
            k_12 = x_222(x1,x2,alpha,f1, f2, f3, d1, d2, d3, H)
        
            k_21 = x_111( (x1 + h_2 * k_11), (x2 + h_2 * k_12),alpha,f1, f2, f3, d1, d2, d3, H)
            k_22 = x_222( (x1 + h_2 * k_11), (x2 + h_2 * k_12),alpha,f1, f2, f3, d1, d2, d3, H)
                
            k_31 = x_111( (x1 + h_2 * k_21), (x2 + h_2 * k_22),alpha,f1, f2, f3, d1, d2, d3, H)
            k_32 = x_222( (x1 + h_2 * k_21), (x2 + h_2 * k_22),alpha,f1, f2, f3, d1, d2, d3, H)
                
            k_41 = x_111( (x1 + h_int * k_31), (x2 + h_int * k_32),alpha,f1, f2, f3, d1, d2, d3, H)
            k_42 = x_222( (x1 + h_int * k_31), (x2 + h_int * k_32),alpha,f1, f2, f3, d1, d2, d3, H)
    
            x1 += h_6 * (k_11 + 2 * k_21 + 2 * k_31 + k_41)
            x2 += h_6 * (k_12 + 2 * k_22 + 2 * k_32 + k_42)
            x3 = 1 - x1 - x2
                
            #print(x1,x2,x3)
                    
            t += h_int
            
        if (save_dynamics_equal_physiological_rates or save_dynamics_different_physiological_rates) and (f is not None):
            f.write(str(t) + " " + str(x1) + " " + str(x2) + " " + str(x3) + "\n")
            
        data[0, i]= t
        data[1, i] = x1
        data[2, i] = x2
        data[3, i] = x3
        
        # BREAK LOOP IF EXPLOTES       
        if x1 < 1e-12 or x2 < 1e-12 or x3 < 1e-12:
            numerical_extinction = True
            break

            
    if f is not None:
        f.close()
  
    return data, numerical_extinction


def rk4_alpha(nsteps, h_int, dt, x10, x20, x30, ALPHAS, f1, f2, f3, d1, d2, d3, H):
    AC = np.nan 
    for alpha in ALPHAS:
        #print(alpha)
        
        data, numerical_extinction = rk4(nsteps, h_int, dt, x10, x20, x30, alpha, f1, f2, f3, d1, d2, d3, H)
        
        
        if numerical_extinction == True:
            continue
        
        if np.isnan(data).any():
            continue

        final_x1, final_x2, final_x3 = data[1, -1], data[2, -1], data[3, -1]
        if final_x1 < 1e-3 or final_x2 < 1e-3 or final_x3 < 1e-3:
            continue
        
        Af = data[1, nsteps-int(nsteps/5):-1]  # 20% of latest points
        Ai = data[1, int(nsteps/5) : int(nsteps/5) + int(nsteps/5)]  # 20% to 40% of earliest points
        
        Af = (max(Af) - min(Af)) / 2  # amplitude of oscillations
        Ai = (max(Ai) - min(Ai)) / 2
        #print(Af, Ai)
        
        amplitude_decrease = (Af + 0.005) - Ai
        std_last = np.std(data[1, nsteps - int(nsteps / 10): -1])
        
        if amplitude_decrease < 0 or std_last < 0.01:  # final amplitude is smaller than initial or last 10% points are cte --> transition occurs
            AC = alpha
            break
    
    return AC

def f(std,mean):
    return np.random.normal(mean,std)
def d(std,mean):
    return np.random.normal(mean,std)

def run_simulation_with_gaussian_sampling(nsteps, h_int, dt, ALPHAS, num_trials, std_list, mean_f_d, x10, x20, x30, H):
    results = []
    for std in std_list:
        
        num_trials_done = 0  # Track the number of accepted trials for this std
        
        while num_trials_done < num_trials:
            
            # Generate f1, f2, f3, d1, d2, d3 from Gaussian distributions
            f1 = f(std, mean_f_d)
            f2 = f(std, mean_f_d)
            f3 = f(std, mean_f_d)
            d1 = d(std, mean_f_d)
            d2 = d(std, mean_f_d)
            d3 = d(std, mean_f_d)
            #print(f1,f2,f3,d1,d2,d3)

            # Normalize f and d by the maximum value to avoid numerical issues
            max_fd = max(f1, f2, f3, d1, d2, d3)
            f1_norm = f1 / max_fd
            f2_norm = f2 / max_fd
            f3_norm = f3 / max_fd
            d1_norm = d1 / max_fd
            d2_norm = d2 / max_fd
            d3_norm = d3 / max_fd
            
            if any(value <= 0 for value in [f1, f2, f3, d1, d2, d3]):
                # Skip this trial and continue generating new values
                continue

            # Mean rate for adjusting nsteps if necessary
            mean_rate = (f1_norm + f2_norm + f3_norm + d1_norm + d2_norm + d3_norm) / 6
            adjusted_nsteps = int(nsteps / mean_rate)

            alpha_c = rk4_alpha(adjusted_nsteps, h_int, dt, x10, x20, x30, ALPHAS,
                                f1_norm, f2_norm, f3_norm, d1_norm, d2_norm, d3_norm, H)

            p_c_i = 1 if not np.isnan(alpha_c) else 0

            # Calculate std of f and d
            std_f = np.std([f1, f2, f3])
            std_d = np.std([d1, d2, d3])

            # Calculate ratios_C and their std
            ratios_C = [f1 / d1, f2 / d2, f3 / d3]
            std_C = np.std(ratios_C)

            results.append({
                'std_gaussian': std,
                'f': [f1, f2, f3],
                'std_f': std_f,
                'd': [d1, d2, d3],
                'std_d': std_d,
                'ratios_C': ratios_C,
                'std_ratios_C': std_C,
                'alpha_c': alpha_c,
                'p_c_i': p_c_i # 1 if there is coexistence, otherwise 0
            })
            num_trials_done += 1
            
    return results

def run_simulation_with_real_data(plant_data, nsteps, h_int, dt, ALPHAS, num_trials, x10, x20, x30, H):
    results = []
    used_combinations = set()
    while len(results) < num_trials:

        indices = tuple(sorted(random.sample(range(len(plant_data)), 3)))
        #print(indices)
        if indices in used_combinations:
            continue
        used_combinations.add(indices)
        species_data_1 = plant_data.iloc[indices[0]]
        species_data_2 = plant_data.iloc[indices[1]]
        species_data_3 = plant_data.iloc[indices[2]]
        #print(species_data_1)
        #print(species_data_2)
        #print(species_data_3)
        max_fd = max(species_data_1[2], species_data_2[2], species_data_3[2],species_data_1[3], species_data_2[3], species_data_3[3])
        F1, D1 = species_data_1[2], species_data_1[3]
        F2, D2 = species_data_2[2], species_data_2[3]
        F3, D3 = species_data_3[2], species_data_3[3]
        f1, d1 = species_data_1[2] / max_fd, species_data_1[3] / max_fd 
        f2, d2 = species_data_2[2] / max_fd, species_data_2[3] / max_fd
        f3, d3 = species_data_3[2] / max_fd, species_data_3[3] / max_fd
        #print(f1, d1, f2, d2, f3, d3)
        
        mean_rate = (f1 + f2 + f3 + d1 + d2 + d3) / 6
        #print('mean rate', mean_rate)
        
        adjusted_nsteps = int(nsteps / mean_rate)
        #print('adjusted nsteps', adjusted_nsteps)
        
        alpha_c = rk4_alpha(adjusted_nsteps, h_int, dt, x10, x20, x30, ALPHAS, f1, f2, f3, d1, d2, d3, H)
        #print(alpha_c)
        p_c_i = 1 if not np.isnan(alpha_c) else 0
        
        # Calculate std of f and d separately
        std_f = np.std([F1, F2, F3])
        std_d = np.std([D1, D2, D3])
        
        # Calculate std and proportion of each pair (f_i, d_i)
        ratios_C = [F1/D1, F2/D2, F3/D3]
        std_C = np.std([F1 / D1, F2 / D2, F3 / D3])
        
        results.append({
            'indices': indices,
            'f': [F1, F2, F3],
            'std_f': std_f,
            'd': [D1, D2, D3],
            'std_d': std_d,
            'ratios_C': ratios_C,           
            'std_ratios_C': std_C,
            'alpha_c': alpha_c,
            'p_c_i': p_c_i # 1 if there is coexistence, otherwise 0
        })
        
    return results
```

```python
import time
import numpy as np
import pandas as pd

def run_benchmark(nsteps, num_trials_per_std, n_alphas, label, seed=123):
    global save_dynamics_equal_physiological_rates, save_dynamics_different_physiological_rates
    save_dynamics_equal_physiological_rates = False
    save_dynamics_different_physiological_rates = False

    h_int = 1
    dt = 1
    ALPHAS = np.linspace(0, 1.0, n_alphas)

    # stds del benchmark (pon los que quieras)
    std_list = [0.01, 0.1, 0.5]
    mean_f_d = 1.0

    # IC fija
    x10, x20, x30 = 0.32, 0.21, 0.47

    # H fija (para que SOLO cambie nsteps)
    H = np.array([[0.5, 0.34, 0.76],
                  [0.66,0.5, 0.25],
                  [0.24,0.75,0.5]], dtype=float)

    # OJO: para reproducibilidad del muestreo, fuerza semilla aquí
    np.random.seed(seed)

    t0 = time.time()
    results = run_simulation_with_gaussian_sampling(
        nsteps=nsteps,
        h_int=h_int,
        dt=dt,
        ALPHAS=ALPHAS,
        num_trials=num_trials_per_std,
        std_list=std_list,
        mean_f_d=mean_f_d,
        x10=x10, x20=x20, x30=x30,
        H=H
    )
    t1 = time.time()

    df = pd.DataFrame(results)

    print(f"\n=== {label} ===")
    print("seconds:", t1 - t0)
    print("rows:", len(df))

    # resumen por std
    for std in std_list:
        dfi = df[df["std_gaussian"] == std]
        alpha_mean = dfi["alpha_c"].dropna().mean()
        pcoex = dfi["p_c_i"].mean()
        print(f"\n  std = {std}")
        print("    alpha_c mean (non-NaN):", alpha_mean)
        print("    coexist fraction:", pcoex)

    return df, (t1 - t0)
```

```python
import time
import numpy as np
import pandas as pd

def run_benchmark(nsteps, num_trials_per_std, n_alphas, label):
    np.random.seed(123)

    global save_dynamics_different_physiological_rates
    global save_dynamics_equal_physiological_rates
    global H

    save_dynamics_different_physiological_rates = False
    save_dynamics_equal_physiological_rates = False

    h_int = 1
    dt = 1
    mean_f_d = 1.0
    x10, x20, x30 = 0.32, 0.11, 0.57
    ALPHAS = np.linspace(0, 1.0, n_alphas)
    std_list = [0.01, 0.1, 0.5]

    H = [[0.5, 0.34, 0.76],
         [0.66, 0.5, 0.25],
         [0.24, 0.75, 0.5]]

    t0 = time.perf_counter()

    results = run_simulation_with_gaussian_sampling(
        nsteps, h_int, dt, ALPHAS,
        num_trials_per_std, std_list, mean_f_d , x10, x20, x30, H
    )

    t1 = time.perf_counter()

    df = pd.DataFrame(results)

    print("\n===", label, "===")
    print("seconds:", t1 - t0)
    print("rows:", len(df))

    # ---- grouped stats per std ----
    grouped = df.groupby("std_gaussian")

    for std_value, subdf in grouped:
        alpha_mean = subdf["alpha_c"].dropna().mean()
        coexist_fraction = subdf["p_c_i"].mean()

        print(f"\n  std = {std_value}")
        print("    alpha_c mean (non-NaN):", alpha_mean)
        print("    coexist fraction:", coexist_fraction)

    return df, (t1 - t0)
```

```python
import numpy as np
import matplotlib.pyplot as plt

def plot_dynamics_over_alphas(
    H,
    alphas,
    nsteps=1000,
    h_int=1,
    dt=1,
    x0=(0.32, 0.21, 0.47),
    f=(1.0, 1.0, 1.0),
    d=(1.0, 1.0, 1.0),
    ylim=(0, 1),
):
    x10, x20, x30 = x0
    f1, f2, f3 = f
    d1, d2, d3 = d

    for alpha in alphas:
        data, numerical_extinction = rk4(
            nsteps, h_int, dt,
            x10, x20, x30,
            alpha, f1, f2, f3, d1, d2, d3,
            H
        )

        # correct indexing
        t  = data[0, :]
        x1 = data[1, :]
        x2 = data[2, :]
        x3 = data[3, :]

        plt.figure()
        plt.plot(t, x1, 'm', label='x1')
        plt.plot(t, x2, 'b', label='x2')
        plt.plot(t, x3, 'y', label='x3')
        plt.ylim(*ylim)
        plt.yticks([0, 0.5, 1])
        plt.tick_params(axis='x', labelsize=12)
        plt.tick_params(axis='y', labelsize=12)
        plt.xlabel(r"$t$", fontsize=14)
        plt.ylabel(r"$x_i(t)$", fontsize=14)
        plt.title(f"alpha={alpha:.3f} | extinction={numerical_extinction}")
        plt.legend()
        plt.show()
```

```python
def alpha_c_vs_nsteps(
    H,
    nsteps_list,
    ALPHAS,
    h_int=1,
    dt=1,
    x0=(0.32, 0.21, 0.47),
    f=(1.0, 1.0, 1.0),
    d=(1.0, 1.0, 1.0),
):
    x10, x20, x30 = x0
    f1, f2, f3 = f
    d1, d2, d3 = d

    out = []
    for nsteps in nsteps_list:
        ac = rk4_alpha(
            nsteps, h_int, dt,
            x10, x20, x30,
            ALPHAS, f1, f2, f3, d1, d2, d3,
            H
        )
        out.append((nsteps, ac))
    return out

# Example usage
ALPHAS = np.linspace(0, 1.0, 81)[1:]  # if you want to skip alpha=0 globally here
nsteps_list = [250, 500, 1000, 2000, 3000, 5000]

res = alpha_c_vs_nsteps(H_fixed, nsteps_list, ALPHAS, f=(1.2, 0.9, 1.04), d=(1.3, 0.87, 0.9))
print(res)
```

```python
import numpy as np

def robustness_check_nsteps(
    H_list,
    std=0.5,
    mean_f_d=1.0,
    n_trials=50,
    nsteps_small=500,
    nsteps_ref=5000,
    ALPHAS=None,
    h_int=1,
    dt=1,
    x0=(0.32, 0.21, 0.47),
    seed=123
):
    rng = np.random.default_rng(seed)
    if ALPHAS is None:
        ALPHAS = np.linspace(0, 1.0, 81)  # use your production grid

    x10, x20, x30 = x0
    step = ALPHAS[1] - ALPHAS[0]

    summary = []
    for h_idx, H in enumerate(H_list):
        n_ok = 0
        n_ok_1step = 0
        n_both_nan = 0
        n_any_nan = 0

        for _ in range(n_trials):
            # sample f,d (reject nonpositive)
            while True:
                f = rng.normal(mean_f_d, std, size=3)
                d = rng.normal(mean_f_d, std, size=3)
                if np.all(f > 0) and np.all(d > 0):
                    break

            # normalize by max to match your pipeline
            max_fd = max(np.max(f), np.max(d))
            f1,f2,f3 = f / max_fd
            d1,d2,d3 = d / max_fd

            ac_small = rk4_alpha(nsteps_small, h_int, dt, x10, x20, x30, ALPHAS,
                                 f1,f2,f3,d1,d2,d3, H)
            ac_ref   = rk4_alpha(nsteps_ref, h_int, dt, x10, x20, x30, ALPHAS,
                                 f1,f2,f3,d1,d2,d3, H)

            if np.isnan(ac_small) and np.isnan(ac_ref):
                n_both_nan += 1
                n_ok += 1
                n_ok_1step += 1
                continue
            if np.isnan(ac_small) or np.isnan(ac_ref):
                n_any_nan += 1
                continue

            if ac_small == ac_ref:
                n_ok += 1
            if abs(ac_small - ac_ref) <= step:
                n_ok_1step += 1

        summary.append({
            "H_index": h_idx,
            "trials": n_trials,
            "exact_match_frac": n_ok / n_trials,
            "within_1step_frac": n_ok_1step / n_trials,
            "both_nan_frac": n_both_nan / n_trials,
            "any_nan_frac": n_any_nan / n_trials
        })

    return summary
```

```python
H_fixed = np.array([[0.5, 0.34, 0.76],
                    [0.66, 0.5, 0.25],
                    [0.24, 0.75, 0.5]])

rng = np.random.default_rng(0)
H_rand2 = random_intransitive_H_3(rng=rng)

ALPHAS = np.linspace(0, 1.0, 81)

summary = robustness_check_nsteps(
    H_list=[H_fixed, H_rand2],
    std=0.5,
    n_trials=50,
    nsteps_small=500,
    nsteps_ref=5000,
    ALPHAS=ALPHAS
)

for s in summary:
    print(s)
```

```python
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

os.chdir(r"C:\Users\mduran\Desktop\PAPERS\HOIs competitive communities\HOIs_dynamics")
csv_path = "gaussian_multiH_all.csv"
df = pd.read_csv(csv_path)

for c in ["std_gaussian", "alpha_c", "p_c_i", "std_f", "std_d", "std_ratios_C", "h_id"]:
    if c in df.columns:
        df[c] = pd.to_numeric(df[c], errors="coerce")

if "p_c_i" not in df.columns and "alpha_c" in df.columns:
    df["p_c_i"] = (~df["alpha_c"].isna()).astype(int)
df["p_c_i"] = pd.to_numeric(df["p_c_i"], errors="coerce").round().clip(0, 1)

df_std = df.dropna(subset=["std_gaussian"]).copy()
df_ratio = df.dropna(subset=["std_ratios_C"]).copy()

def log_binned_summary(d, x_col, n_bins=12, by_h=True, h_col="h_id",
                       alpha_col="alpha_c", coex_col="p_c_i"):
    d = d.copy()
    d = d.dropna(subset=[x_col])
    d = d[np.isfinite(d[x_col].values)]
    d = d[d[x_col] > 0]
    if d.empty:
        return pd.DataFrame()

    xmin, xmax = float(d[x_col].min()), float(d[x_col].max())
    if (not np.isfinite(xmin)) or (not np.isfinite(xmax)) or xmin <= 0 or xmax <= 0 or xmin == xmax:
        d["x_bin"] = "all"
        d["x_center"] = float(d[x_col].median())
    else:
        edges = np.logspace(np.log10(xmin), np.log10(xmax), n_bins + 1)
        d["x_bin"] = pd.cut(d[x_col], bins=edges, include_lowest=True)
        d = d.dropna(subset=["x_bin"])
        if d.empty:
            return pd.DataFrame()

        intervals = d["x_bin"].cat.categories
        centers = np.sqrt(intervals.left.to_numpy() * intervals.right.to_numpy())
        center_map = {intervals[i]: float(centers[i]) for i in range(len(intervals))}
        d["x_center"] = d["x_bin"].map(center_map)

    d["x_center"] = pd.to_numeric(d["x_center"], errors="coerce")
    d[alpha_col] = pd.to_numeric(d[alpha_col], errors="coerce")
    d["p01"] = pd.to_numeric(d[coex_col], errors="coerce").round().clip(0, 1)

    group_cols = ([h_col, "x_bin"] if by_h else ["x_bin"])

    out = (
        d.groupby(group_cols, dropna=False)
         .agg(
            x_center=("x_center", "first"),
            alpha_mean=(alpha_col, "mean"),
            p_mean=("p01", "mean"),
         )
         .reset_index()
    )

    sort_cols = ([h_col, "x_center"] if by_h else ["x_center"])
    return out.sort_values(sort_cols).reset_index(drop=True)

def plot_mean(summary, x_col, y_col, title, xlabel, ylabel, by_h=True, h_col="h_id", logx=True):
    plt.figure()
    if summary is None or summary.empty:
        print(f"[WARN] Empty summary: {title}")
        return

    if by_h:
        for hid, sub in summary.groupby(h_col):
            sub = sub.sort_values(x_col)
            x = pd.to_numeric(sub[x_col], errors="coerce").to_numpy(dtype=float)
            y = pd.to_numeric(sub[y_col], errors="coerce").to_numpy(dtype=float)
            ok = np.isfinite(x) & np.isfinite(y)
            if np.any(ok):
                plt.plot(x[ok], y[ok], marker="o", linewidth=1, label=f"H {hid}")
        plt.legend(ncol=2, fontsize=8)
    else:
        sub = summary.sort_values(x_col)
        x = pd.to_numeric(sub[x_col], errors="coerce").to_numpy(dtype=float)
        y = pd.to_numeric(sub[y_col], errors="coerce").to_numpy(dtype=float)
        ok = np.isfinite(x) & np.isfinite(y)
        if np.any(ok):
            plt.plot(x[ok], y[ok], marker="o", linewidth=2)

    plt.title(title)
    plt.xlabel(xlabel)
    if logx:
        plt.xscale("log")
    plt.ylabel(ylabel)
    plt.grid(True, alpha=0.3)
    plt.tight_layout()
    plt.show()

# ---- Summaries ----
N_BINS = 12
sum_std_perH  = log_binned_summary(df_std,   x_col="std_gaussian", n_bins=N_BINS, by_h=True)
sum_std_allH  = log_binned_summary(df_std,   x_col="std_gaussian", n_bins=N_BINS, by_h=False)
sum_rat_perH  = log_binned_summary(df_ratio, x_col="std_ratios_C",  n_bins=N_BINS, by_h=True)
sum_rat_allH  = log_binned_summary(df_ratio, x_col="std_ratios_C",  n_bins=N_BINS, by_h=False)

# ---- Fig.3-like ----
plot_mean(sum_std_perH, "x_center", "alpha_mean", "Fig.3-like: alpha_c mean vs std (per H)",
          "std_gaussian [bin centers]", "mean alpha_c", by_h=True, logx=True)
plot_mean(sum_std_allH, "x_center", "alpha_mean", "Fig.3-like: alpha_c mean vs std (all H)",
          "std_gaussian [bin centers]", "mean alpha_c", by_h=False, logx=True)

plot_mean(sum_std_perH, "x_center", "p_mean", "Fig.3-like: p_coex mean vs std (per H)",
          "std_gaussian [bin centers]", "p_coex", by_h=True, logx=True)
plot_mean(sum_std_allH, "x_center", "p_mean", "Fig.3-like: p_coex mean vs std (all H)",
          "std_gaussian [bin centers]", "p_coex", by_h=False, logx=True)

# ---- Fig.4-like ----
plot_mean(sum_rat_perH, "x_center", "alpha_mean", "Fig.4-like: alpha_c mean vs std_ratios_C (per H)",
          "std_ratios_C (= std_ratios_C/std_gaussian) [bin centers]", "mean alpha_c", by_h=True, logx=True)
plot_mean(sum_rat_allH, "x_center", "alpha_mean", "Fig.4-like: alpha_c mean vs std_ratios_C (all H)",
          "std_ratios_C (= std_ratios_C/std_gaussian) [bin centers]", "mean alpha_c", by_h=False, logx=True)

plot_mean(sum_rat_perH, "x_center", "p_mean", "Fig.4-like: p_coex mean vs std_ratios_C (per H)",
          "std_ratios_C (= std_ratios_C/std_gaussian) [bin centers]", "p_coex", by_h=True, logx=True)
plot_mean(sum_rat_allH, "x_center", "p_mean", "Fig.4-like: p_coex mean vs std_ratios_C (all H)",
          "std_ratios_C (= std_ratios_C/std_gaussian) [bin centers]", "p_coex", by_h=False, logx=True)
```

```python
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

os.chdir(r"C:\Users\mduran\Desktop\PAPERS\HOIs competitive communities\HOIs_dynamics")
csv_path = "realdata_multiH_all.csv"
df = pd.read_csv(csv_path)

# --- numeric columns ---
for c in ["alpha_c", "p_c_i", "std_f", "std_d", "std_ratios_C", "h_id"]:
    if c in df.columns:
        df[c] = pd.to_numeric(df[c], errors="coerce")

if "p_c_i" not in df.columns and "alpha_c" in df.columns:
    df["p_c_i"] = (~df["alpha_c"].isna()).astype(int)
df["p_c_i"] = pd.to_numeric(df["p_c_i"], errors="coerce").round().clip(0, 1)

if "plant_type" not in df.columns:
    raise ValueError("Expected column 'plant_type' in realdata_multiH_all.csv")

X_CHOICES = [("std_ratios_C", "std_ratios_C")]

# -----------------------------
# BINNED SUMMARY (same as before but you can keep your full version)
# -----------------------------
def log_binned_summary(d, x_col, n_bins=12, by_h=True, h_col="h_id",
                       alpha_col="alpha_c", coex_col="p_c_i"):
    d = d.copy()
    d = d.dropna(subset=[x_col])
    d = d[np.isfinite(d[x_col].values)]
    d = d[d[x_col] > 0]
    if d.empty:
        return pd.DataFrame()

    xmin, xmax = float(d[x_col].min()), float(d[x_col].max())

    if (not np.isfinite(xmin)) or (not np.isfinite(xmax)) or xmin <= 0 or xmax <= 0 or xmin == xmax:
        d["x_bin"] = "all"
        d["x_center"] = float(d[x_col].median())
    else:
        edges = np.logspace(np.log10(xmin), np.log10(xmax), n_bins + 1)
        d["x_bin"] = pd.cut(d[x_col], bins=edges, include_lowest=True)
        d = d.dropna(subset=["x_bin"])
        if d.empty:
            return pd.DataFrame()

        intervals = d["x_bin"].cat.categories
        centers = np.sqrt(intervals.left.to_numpy() * intervals.right.to_numpy())
        center_map = {intervals[i]: float(centers[i]) for i in range(len(intervals))}
        d["x_center"] = d["x_bin"].map(center_map)

    d["x_center"] = pd.to_numeric(d["x_center"], errors="coerce")
    d[alpha_col] = pd.to_numeric(d[alpha_col], errors="coerce")
    d["p01"] = pd.to_numeric(d[coex_col], errors="coerce").round().clip(0, 1)

    group_cols = ([h_col, "x_bin"] if by_h else ["x_bin"])

    out = (
        d.groupby(group_cols, dropna=False)
         .agg(
            x_center=("x_center", "first"),
            alpha_mean=(alpha_col, "mean"),
            p_mean=("p01", "mean"),
         )
         .reset_index()
    )

    sort_cols = ([h_col, "x_center"] if by_h else ["x_center"])
    return out.sort_values(sort_cols).reset_index(drop=True)

# -----------------------------
# PLOTTING: mean-only
# -----------------------------
def plot_mean(summary, x_col, y_col, title, xlabel, ylabel, by_h=True, h_col="h_id", logx=True):
    plt.figure()
    if summary is None or summary.empty:
        print(f"[WARN] Empty summary: {title}")
        return

    if by_h:
        for hid, sub in summary.groupby(h_col):
            sub = sub.sort_values(x_col)
            x = pd.to_numeric(sub[x_col], errors="coerce").to_numpy(dtype=float)
            y = pd.to_numeric(sub[y_col], errors="coerce").to_numpy(dtype=float)
            ok = np.isfinite(x) & np.isfinite(y)
            if np.any(ok):
                plt.plot(x[ok], y[ok], marker="o", linewidth=1, label=f"H {hid}")
        plt.legend(ncol=2, fontsize=8)
    else:
        sub = summary.sort_values(x_col)
        x = pd.to_numeric(sub[x_col], errors="coerce").to_numpy(dtype=float)
        y = pd.to_numeric(sub[y_col], errors="coerce").to_numpy(dtype=float)
        ok = np.isfinite(x) & np.isfinite(y)
        if np.any(ok):
            plt.plot(x[ok], y[ok], marker="o", linewidth=2)

    plt.title(title)
    plt.xlabel(xlabel)
    if logx:
        plt.xscale("log")
    plt.ylabel(ylabel)
    plt.grid(True, alpha=0.3)
    plt.tight_layout()
    plt.show()

def make_plots_for_plants(dplant, plant_label, x_col, x_label, n_bins=12):
    sum_perH = log_binned_summary(dplant, x_col=x_col, n_bins=n_bins, by_h=True)
    sum_allH = log_binned_summary(dplant, x_col=x_col, n_bins=n_bins, by_h=False)

    # alpha trend
    plot_mean(
        sum_perH, x_col="x_center", y_col="alpha_mean",
        title=f"{plant_label} | alpha_c mean vs {x_label} (binned, per H)",
        xlabel=f"{x_label} [bin centers]", ylabel="mean alpha_c",
        by_h=True, logx=True
    )
    plot_mean(
        sum_allH, x_col="x_center", y_col="alpha_mean",
        title=f"{plant_label} | alpha_c mean vs {x_label} (binned, all H)",
        xlabel=f"{x_label} [bin centers]", ylabel="mean alpha_c",
        by_h=False, logx=True
    )

    # p_coex trend
    plot_mean(
        sum_perH, x_col="x_center", y_col="p_mean",
        title=f"{plant_label} | p_coex mean vs {x_label} (binned, per H)",
        xlabel=f"{x_label} [bin centers]", ylabel="p_coex",
        by_h=True, logx=True
    )
    plot_mean(
        sum_allH, x_col="x_center", y_col="p_mean",
        title=f"{plant_label} | p_coex mean vs {x_label} (binned, all H)",
        xlabel=f"{x_label} [bin centers]", ylabel="p_coex",
        by_h=False, logx=True
    )

# -----------------------------
# RUN
# -----------------------------
N_BINS = 12

for plant_type in sorted(df["plant_type"].dropna().unique()):
    dplant = df[df["plant_type"] == plant_type].copy()

    # Fig.3-like (std_ratios_C only, as you set)
    for x_col, x_label in X_CHOICES:
        make_plots_for_plants(dplant, plant_label=f"Plant = {plant_type}", x_col=x_col, x_label=x_label, n_bins=N_BINS)

    # Fig.4-like (ratio axis)
    make_plots_for_plants(
        dplant, plant_label=f"Plant = {plant_type}",
        x_col="std_ratios_C", x_label="std_ratios_C (= std_ratios_C/(std_f+std_d))",
        n_bins=N_BINS
    )
```
