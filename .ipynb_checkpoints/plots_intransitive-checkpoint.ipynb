```python
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

os.chdir(r"C:\Users\mduran\Desktop\PAPERS\HOIs competitive communities\HOIs_dynamics")
csv_path = "gaussian_multiH_all.csv"
df = pd.read_csv(csv_path)

for c in ["std_gaussian", "alpha_c", "p_c_i", "std_f", "std_d", "std_ratios_C", "H_id"]:
    if c in df.columns:
        df[c] = pd.to_numeric(df[c], errors="coerce")

if "p_c_i" not in df.columns and "alpha_c" in df.columns:
    df["p_c_i"] = (~df["alpha_c"].isna()).astype(int)
df["p_c_i"] = pd.to_numeric(df["p_c_i"], errors="coerce").round().clip(0, 1)

df_std = df.dropna(subset=["std_gaussian"]).copy()
df_ratio = df.dropna(subset=["std_ratios_C"]).copy()

def direct_summary(d, x_col, by_h=True, h_col="H_id",
                   alpha_col="alpha_c", coex_col="p_c_i"):
    d = d.copy()
    d = d.dropna(subset=[x_col])
    d = d[np.isfinite(d[x_col].values)]
    d = d[d[x_col] > 0]

    if d.empty:
        return pd.DataFrame()

    d[alpha_col] = pd.to_numeric(d[alpha_col], errors="coerce")
    d["p01"] = pd.to_numeric(d[coex_col], errors="coerce").round().clip(0, 1)

    group_cols = ([h_col, x_col] if by_h else [x_col])

    out = (
        d.groupby(group_cols)
         .agg(
            alpha_mean=(alpha_col, "mean"),
            p_mean=("p01", "mean"),
         )
         .reset_index()
         .rename(columns={x_col: "x_center"})
    )

    sort_cols = ([h_col, "x_center"] if by_h else ["x_center"])
    return out.sort_values(sort_cols).reset_index(drop=True)

def log_binned_summary(d, x_col, n_bins=12, by_h=True, h_col="H_id",
                       alpha_col="alpha_c", coex_col="p_c_i"):
    d = d.copy()
    d = d.dropna(subset=[x_col])
    d = d[np.isfinite(d[x_col].values)]
    d = d[d[x_col] > 0]
    if d.empty:
        return pd.DataFrame()

    xmin, xmax = float(d[x_col].min()), float(d[x_col].max())
    if (not np.isfinite(xmin)) or (not np.isfinite(xmax)) or xmin <= 0 or xmax <= 0 or xmin == xmax:
        d["x_bin"] = "all"
        d["x_center"] = float(d[x_col].median())
    else:
        edges = np.logspace(np.log10(xmin), np.log10(xmax), n_bins + 1)
        d["x_bin"] = pd.cut(d[x_col], bins=edges, include_lowest=True)
        d = d.dropna(subset=["x_bin"])
        if d.empty:
            return pd.DataFrame()

        intervals = d["x_bin"].cat.categories
        centers = np.sqrt(intervals.left.to_numpy() * intervals.right.to_numpy())
        center_map = {intervals[i]: float(centers[i]) for i in range(len(intervals))}
        d["x_center"] = d["x_bin"].map(center_map)

    d["x_center"] = pd.to_numeric(d["x_center"], errors="coerce")
    d[alpha_col] = pd.to_numeric(d[alpha_col], errors="coerce")
    d["p01"] = pd.to_numeric(d[coex_col], errors="coerce").round().clip(0, 1)

    group_cols = ([h_col, "x_bin"] if by_h else ["x_bin"])

    out = (
        d.groupby(group_cols, dropna=False)
         .agg(
            x_center=("x_center", "first"),
            alpha_mean=(alpha_col, "mean"),
            p_mean=("p01", "mean"),
         )
         .reset_index()
    )

    sort_cols = ([h_col, "x_center"] if by_h else ["x_center"])
    return out.sort_values(sort_cols).reset_index(drop=True)

def plot_mean(summary, x_col, y_col, title, xlabel, ylabel, by_h=True, h_col="H_id", logx=True):
    plt.figure()
    if summary is None or summary.empty:
        print(f"[WARN] Empty summary: {title}")
        return

    if by_h:
        for hid, sub in summary.groupby(h_col):
            sub = sub.sort_values(x_col)
            x = pd.to_numeric(sub[x_col], errors="coerce").to_numpy(dtype=float)
            y = pd.to_numeric(sub[y_col], errors="coerce").to_numpy(dtype=float)
            ok = np.isfinite(x) & np.isfinite(y)
            if np.any(ok):
                plt.plot(x[ok], y[ok], marker="o", linewidth=1, label=f"H {hid}")
        plt.legend(ncol=2, fontsize=8)
    else:
        sub = summary.sort_values(x_col)
        x = pd.to_numeric(sub[x_col], errors="coerce").to_numpy(dtype=float)
        y = pd.to_numeric(sub[y_col], errors="coerce").to_numpy(dtype=float)
        ok = np.isfinite(x) & np.isfinite(y)
        if np.any(ok):
            plt.plot(x[ok], y[ok], marker="o", linewidth=2)

    plt.title(title)
    plt.xlabel(xlabel)
    if logx:
        plt.xscale("log")
    plt.ylabel(ylabel)
    plt.grid(True, alpha=0.3)
    plt.tight_layout()
    plt.show()

# ---- Summaries ----
N_BINS = 15
sum_std_perH  = direct_summary(df_std,   x_col="std_gaussian", by_h=True)
sum_std_allH  = direct_summary(df_std,   x_col="std_gaussian", by_h=False)
sum_rat_perH  = log_binned_summary(df_ratio, x_col="std_ratios_C",  n_bins=N_BINS, by_h=True)
sum_rat_allH  = log_binned_summary(df_ratio, x_col="std_ratios_C",  n_bins=N_BINS, by_h=False)

# ---- Fig.3-like ----
plot_mean(sum_std_perH, "x_center", "alpha_mean", "Fig.3-like: alpha_c mean vs std (per H)",
          "std_gaussian [bin centers]", "mean alpha_c", by_h=True, logx=True)
plot_mean(sum_std_allH, "x_center", "alpha_mean", "Fig.3-like: alpha_c mean vs std (all H)",
          "std_gaussian [bin centers]", "mean alpha_c", by_h=False, logx=True)

plot_mean(sum_std_perH, "x_center", "p_mean", "Fig.3-like: p_coex mean vs std (per H)",
          "std_gaussian [bin centers]", "p_coex", by_h=True, logx=True)
plot_mean(sum_std_allH, "x_center", "p_mean", "Fig.3-like: p_coex mean vs std (all H)",
          "std_gaussian [bin centers]", "p_coex", by_h=False, logx=True)

# ---- Fig.4-like ----
plot_mean(sum_rat_perH, "x_center", "alpha_mean", "Fig.4-like: alpha_c mean vs std_ratios_C (per H)",
          "std_ratios_C [bin centers]", "mean alpha_c", by_h=True, logx=True)
plot_mean(sum_rat_allH, "x_center", "alpha_mean", "Fig.4-like: alpha_c mean vs std_ratios_C (all H)",
          "std_ratios_C [bin centers]", "mean alpha_c", by_h=False, logx=True)

plot_mean(sum_rat_perH, "x_center", "p_mean", "Fig.4-like: p_coex mean vs std_ratios_C (per H)",
          "std_ratios_C [bin centers]", "p_coex", by_h=True, logx=True)
plot_mean(sum_rat_allH, "x_center", "p_mean", "Fig.4-like: p_coex mean vs std_ratios_C (all H)",
          "std_ratios_C [bin centers]", "p_coex", by_h=False, logx=True)
```

```python
import os
import re
import glob
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# ============================================================
# CONFIG
# ============================================================
RESULTS_DIR = r"C:\Users\mduran\Desktop\PAPERS\HOIs competitive communities\HOIs_dynamics\results"

# If your files are named exactly like this in RESULTS_DIR:
#   gaussian_multiH_all_H{h_id}.csv
#   {plant}_realdata_multiH_all_H{h_id}.csv   (plant in: Seagrasses, Trees, LandHerbsSaltMarsh)
#   gaussian_multiH_Hmeta.csv   (optional)
#   realdata_multiH_Hmeta.csv   (optional)
#
# Adjust patterns below if your naming differs.

GAUSS_PATTERN = os.path.join(RESULTS_DIR, "gaussian_multiH_all_H*.csv")
PLANT_PATTERN = os.path.join(RESULTS_DIR, "*_realdata_multiH_all_H*.csv")

META  = os.path.join(RESULTS_DIR, "multiH_Hmeta.csv")

PLANT_TYPES = ["Seagrasses", "Trees", "LandHerbsSaltMarsh"]

# numeric tolerance for comparing H entries across files
H_TOL = 1e-12

# Binning for pooled plots over variance of ratios
N_BINS = 15

# ============================================================
# HELPERS
# ============================================================
def _to_num(df, cols):
    for c in cols:
        if c in df.columns:
            df[c] = pd.to_numeric(df[c], errors="coerce")
    return df

def _extract_hid_from_filename(path):
    # expects "...H{integer}.csv"
    m = re.search(r"_H(\d+)\.csv$", os.path.basename(path))
    return int(m.group(1)) if m else None

def _read_csv_safe(path):
    try:
        return pd.read_csv(path)
    except Exception as e:
        print(f"[ERROR] Failed to read {path}: {e}")
        return None

def _h_signature_from_row(row):
    # robust signature from stored H entries + seed
    # (seed helps uniqueness if any rounding issues)
    sig = (
        float(row.get("H01", np.nan)),
        float(row.get("H02", np.nan)),
        float(row.get("H12", np.nan)),
        int(row.get("seed_H", -1)) if pd.notna(row.get("seed_H", np.nan)) else -1
    )
    return sig

def _h_signature_from_df(df):
    if df is None or df.empty:
        return None
    cols = ["H01","H02","H12","seed_H"]
    for c in cols:
        if c not in df.columns:
            return None
    row0 = df.iloc[0]
    return _h_signature_from_row(row0)

def ensure_pc_column(df):
    # p_c should be 1 when alpha_c exists (non-NaN), else 0
    if "p_c_i" not in df.columns and "alpha_c" in df.columns:
        df["p_c_i"] = (~df["alpha_c"].isna()).astype(int)
    df["p_c_i"] = pd.to_numeric(df["p_c_i"], errors="coerce").round().clip(0, 1)
    return df

def log_binned_summary(d, x_col, n_bins=12, by_h=True, h_col="H_id",
                       alpha_col="alpha_c", pc_col="p_c_i"):
    d = d.copy()
    d = d.dropna(subset=[x_col])
    d = d[np.isfinite(d[x_col].values)]
    d = d[d[x_col] > 0]
    if d.empty:
        return pd.DataFrame()

    xmin, xmax = float(d[x_col].min()), float(d[x_col].max())
    if (not np.isfinite(xmin)) or (not np.isfinite(xmax)) or xmin <= 0 or xmax <= 0 or xmin == xmax:
        d["x_bin"] = "all"
        d["x_center"] = float(d[x_col].median())
    else:
        edges = np.logspace(np.log10(xmin), np.log10(xmax), n_bins + 1)
        d["x_bin"] = pd.cut(d[x_col], bins=edges, include_lowest=True)
        d = d.dropna(subset=["x_bin"])
        if d.empty:
            return pd.DataFrame()

        intervals = d["x_bin"].cat.categories
        centers = np.sqrt(intervals.left.to_numpy() * intervals.right.to_numpy())
        center_map = {intervals[i]: float(centers[i]) for i in range(len(intervals))}
        d["x_center"] = d["x_bin"].map(center_map)

    d["x_center"] = pd.to_numeric(d["x_center"], errors="coerce")
    d[alpha_col] = pd.to_numeric(d[alpha_col], errors="coerce")
    d["p01"] = pd.to_numeric(d[pc_col], errors="coerce").round().clip(0, 1)

    group_cols = ([h_col, "x_bin"] if by_h else ["x_bin"])

    out = (
        d.groupby(group_cols, dropna=False)
         .agg(
            x_center=("x_center", "first"),
            alpha_mean=(alpha_col, "mean"),
            p_mean=("p01", "mean"),
            n=("p01", "size"),
         )
         .reset_index()
    )

    sort_cols = ([h_col, "x_center"] if by_h else ["x_center"])
    return out.sort_values(sort_cols).reset_index(drop=True)

def plot_mean(summary, x_col, y_col, title, xlabel, ylabel, by_h=True, h_col="H_id", logx=True):
    plt.figure()
    if summary is None or summary.empty:
        print(f"[WARN] Empty summary: {title}")
        return

    if by_h:
        for hid, sub in summary.groupby(h_col):
            sub = sub.sort_values(x_col)
            x = pd.to_numeric(sub[x_col], errors="coerce").to_numpy(dtype=float)
            y = pd.to_numeric(sub[y_col], errors="coerce").to_numpy(dtype=float)
            ok = np.isfinite(x) & np.isfinite(y)
            if np.any(ok):
                plt.plot(x[ok], y[ok], marker="o", linewidth=1, label=f"H {hid}")
        plt.legend(ncol=2, fontsize=8)
    else:
        sub = summary.sort_values(x_col)
        x = pd.to_numeric(sub[x_col], errors="coerce").to_numpy(dtype=float)
        y = pd.to_numeric(sub[y_col], errors="coerce").to_numpy(dtype=float)
        ok = np.isfinite(x) & np.isfinite(y)
        if np.any(ok):
            plt.plot(x[ok], y[ok], marker="o", linewidth=2)

    plt.title(title)
    plt.xlabel(xlabel)
    if logx:
        plt.xscale("log")
    plt.ylabel(ylabel)
    plt.grid(True, alpha=0.3)
    plt.tight_layout()
    plt.show()

H_TOL = 1e-12  # tolerancia numérica

def _compare_meta_entries(metaA, metaB, tol=1e-12):
    """
    Compare two meta dict-like objects containing:
    H01, H02, H12, seed_H
    """
    # compare seed exactly
    if "seed_H" in metaA and "seed_H" in metaB:
        if int(metaA["seed_H"]) != int(metaB["seed_H"]):
            return False

    # compare H entries with tolerance
    for k in ["H01", "H02", "H12"]:
        if k in metaA and k in metaB:
            a = float(metaA[k])
            b = float(metaB[k])
            if not np.isfinite(a) or not np.isfinite(b):
                return False
            if abs(a - b) > tol:
                return False

    return True
```

```python
# ============================================================
# (i) CHECKS: H consistency across gaussian, plants, meta
# ============================================================
gauss_files = sorted(glob.glob(GAUSS_PATTERN))
plant_files = sorted(glob.glob(PLANT_PATTERN))

print(f"[INFO] Found gaussian files: {len(gauss_files)}")
print(f"[INFO] Found plant files:   {len(plant_files)}")

# Build dict: H_id -> signature from gaussian file
gauss_sig = {}
gauss_meta_from_file = {}

for fp in gauss_files:
    hid = _extract_hid_from_filename(fp)
    df = _read_csv_safe(fp)
    if df is None or df.empty:
        print(f"[WARN] gaussian file empty: {fp}")
        continue
    df = _to_num(df, ["H_id","seed_H","H01","H02","H12","alpha_c","p_c_i","std_ratios_C"])
    sig = _h_signature_from_df(df)
    if sig is None:
        print(f"[WARN] missing H columns in gaussian: {fp}")
        continue
    gauss_sig[hid] = sig
    gauss_meta_from_file[hid] = {
        "H_id": hid,
        "seed_H": int(df["seed_H"].iloc[0]) if "seed_H" in df.columns else -1,
        "H01": float(df["H01"].iloc[0]),
        "H02": float(df["H02"].iloc[0]),
        "H12": float(df["H12"].iloc[0]),
        "intransitive": bool(df["intransitive"].iloc[0]) if "intransitive" in df.columns else None,
    }

# Build dict: (plant_type, H_id) -> signature from plant file
plant_sig = {}
plant_meta_from_file = {}

for fp in plant_files:
    hid = _extract_hid_from_filename(fp)
    df = _read_csv_safe(fp)
    if df is None or df.empty:
        print(f"[WARN] plant file empty: {fp}")
        continue
    df = _to_num(df, ["H_id","seed_H","H01","H02","H12","alpha_c","p_c_i","std_ratios_C"])
    sig = _h_signature_from_df(df)
    if sig is None:
        print(f"[WARN] missing H columns in plant: {fp}")
        continue

    # plant type inferred from filename prefix
    # expects "{plant}_realdata_multiH_all_H{hid}.csv"
    base = os.path.basename(fp)
    plant_type = base.split("_realdata_")[0]
    plant_sig[(plant_type, hid)] = sig
    plant_meta_from_file[(plant_type, hid)] = {
        "plant_type": plant_type,
        "H_id": hid,
        "seed_H": int(df["seed_H"].iloc[0]) if "seed_H" in df.columns else -1,
        "H01": float(df["H01"].iloc[0]),
        "H02": float(df["H02"].iloc[0]),
        "H12": float(df["H12"].iloc[0]),
        "intransitive": bool(df["intransitive"].iloc[0]) if "intransitive" in df.columns else None,
    }

# Load meta files if they exist
gauss_meta_df = _read_csv_safe(META) if os.path.exists(META) else None
real_meta_df  = _read_csv_safe(META)  if os.path.exists(META)  else None

if gauss_meta_df is not None:
    gauss_meta_df = _to_num(gauss_meta_df, ["H_id","seed_H","H01","H02","H12"])
if real_meta_df is not None:
    real_meta_df  = _to_num(real_meta_df,  ["H_id","seed_H","H01","H02","H12"])
```

```python
print("\n==================== (i) CONSISTENCY CHECKS ====================")


# A) Gaussian vs plants
all_hids = sorted(set(gauss_sig.keys()) | set(h for (_, h) in plant_sig.keys()))
bad = 0
for hid in all_hids:
    if hid not in gauss_sig:
        print(f"[WARN] H{hid}: missing gaussian file")
        continue

    gsig = gauss_sig[hid]
    for pt in PLANT_TYPES:
        if (pt, hid) not in plant_sig:
            print(f"[WARN] H{hid}: missing plant file for {pt}")
            continue
        psig = plant_sig[(pt, hid)]

        # compare (H01,H02,H12,seed)
        ok = (abs(gsig[0]-psig[0])<=H_TOL and abs(gsig[1]-psig[1])<=H_TOL and abs(gsig[2]-psig[2])<=H_TOL and gsig[3]==psig[3])
        if not ok:
            bad += 1
            print(f"[MISMATCH] H{hid} gaussian vs {pt}:")
            print(f"  gaussian sig: {gsig}")
            print(f"  plant   sig: {psig}")

if bad == 0:
    print("[OK] All H signatures match between gaussian and plant files (for available files).")
```

```python
# ============================================================
# LOAD DATA FOR PLOTTING
# ============================================================
# Gaussian pooled across H
gauss_dfs = []
for fp in gauss_files:
    df = _read_csv_safe(fp)
    if df is None or df.empty:
        continue
    df = _to_num(df, ["H_id","alpha_c","p_c_i","std_ratios_C"])
    df = ensure_pc_column(df)
    if "H_id" not in df.columns:
        df["H_id"] = _extract_hid_from_filename(fp)
    gauss_dfs.append(df)

df_gauss = pd.concat(gauss_dfs, ignore_index=True) if gauss_dfs else pd.DataFrame()

# Plants pooled across plant types + H
plant_dfs = []
for fp in plant_files:
    df = _read_csv_safe(fp)
    if df is None or df.empty:
        continue
    df = _to_num(df, ["H_id","alpha_c","p_c_i","std_ratios_C"])
    df = ensure_pc_column(df)

    if "H_id" not in df.columns:
        df["H_id"] = _extract_hid_from_filename(fp)

    # infer plant_type if missing
    if "plant_type" not in df.columns:
        base = os.path.basename(fp)
        df["plant_type"] = base.split("_realdata_")[0]

    plant_dfs.append(df)

df_plants = pd.concat(plant_dfs, ignore_index=True) if plant_dfs else pd.DataFrame()

print("\n==================== DATA LOADED ====================")
print(f"[INFO] gaussian rows: {len(df_gauss)}")
print(f"[INFO] plants rows:   {len(df_plants)}")
```

```python
import os
import glob
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# =========================
# CONFIG
# =========================
BASE_DIR = r"C:\Users\mduran\Desktop\PAPERS\HOIs competitive communities\HOIs_dynamics\results"
LOGX = True
N_BINS = 15
H_TOL = 1e-12  # tolerance to compare stored H entries

PLANTS = ["Seagrasses", "Trees", "LandHerbsSaltMarsh"]


# =========================
# HELPERS: IO + CLEANING
# =========================
def _safe_read_csv(path):
    try:
        return pd.read_csv(path)
    except Exception as e:
        print(f"[WARN] Could not read {path}: {e}")
        return None


def _coerce_numeric(df, cols):
    for c in cols:
        if c in df.columns:
            df[c] = pd.to_numeric(df[c], errors="coerce")
    return df


def _ensure_pc(df):
    # p_c_i sometimes exists; if not, infer from alpha_c
    if "p_c_i" not in df.columns and "alpha_c" in df.columns:
        df["p_c_i"] = (~pd.to_numeric(df["alpha_c"], errors="coerce").isna()).astype(int)
    if "p_c_i" in df.columns:
        df["p_c_i"] = pd.to_numeric(df["p_c_i"], errors="coerce").round().clip(0, 1)
    return df


def _parse_hid_from_filename(fname):
    # expects ..._H{hid}.csv
    base = os.path.basename(fname)
    # find last occurrence of "H" followed by digits
    import re
    m = re.search(r"_H(\d+)\.csv$", base)
    if m:
        return int(m.group(1))
    return None


def load_gaussian_files(base_dir):
    # gaussian_multiH_all_H{h}.csv
    pat = os.path.join(base_dir, "gaussian_multiH_all_H*.csv")
    files = sorted(glob.glob(pat))
    if not files:
        print(f"[WARN] No gaussian files found matching: {pat}")
        return pd.DataFrame()

    frames = []
    for fp in files:
        df = _safe_read_csv(fp)
        if df is None or df.empty:
            continue
        hid = _parse_hid_from_filename(fp)
        if "H_id" not in df.columns and hid is not None:
            df["H_id"] = hid
        df["dataset"] = "gaussian"
        frames.append(df)

    out = pd.concat(frames, ignore_index=True) if frames else pd.DataFrame()
    out = _coerce_numeric(out, ["H_id", "alpha_c", "p_c_i", "std_ratios_C", "std_gaussian", "std_f", "std_d"])
    out = _ensure_pc(out)
    return out


def load_plants_files(base_dir, plant_types):
    # {plant}_realdata_multiH_all_H{h}.csv
    frames = []
    for plant in plant_types:
        pat = os.path.join(base_dir, f"{plant}_realdata_multiH_all_H*.csv")
        files = sorted(glob.glob(pat))
        if not files:
            print(f"[WARN] No plant files for {plant} matching: {pat}")
            continue

        for fp in files:
            df = _safe_read_csv(fp)
            if df is None or df.empty:
                continue
            hid = _parse_hid_from_filename(fp)
            if "H_id" not in df.columns and hid is not None:
                df["H_id"] = hid
            if "plant_type" not in df.columns:
                df["plant_type"] = plant
            df["dataset"] = "plants"
            frames.append(df)

    out = pd.concat(frames, ignore_index=True) if frames else pd.DataFrame()
    out = _coerce_numeric(out, ["H_id", "alpha_c", "p_c_i", "std_ratios_C", "std_f", "std_d"])
    out = _ensure_pc(out)
    return out


def load_meta_files(base_dir):
    g_meta_path = os.path.join(base_dir, "gaussian_multiH_Hmeta.csv")
    p_meta_path = os.path.join(base_dir, "realdata_multiH_Hmeta.csv")

    g_meta = _safe_read_csv(g_meta_path)
    p_meta = _safe_read_csv(p_meta_path)

    if g_meta is None:
        g_meta = pd.DataFrame()
    if p_meta is None:
        p_meta = pd.DataFrame()

    for df in (g_meta, p_meta):
        _coerce_numeric(df, ["H_id", "seed_H", "H01", "H02", "H12"])
        if "intransitive" in df.columns:
            # allow True/False or strings
            if df["intransitive"].dtype == object:
                df["intransitive"] = df["intransitive"].astype(str).str.lower().isin(["true", "1", "yes"])
            else:
                df["intransitive"] = df["intransitive"].astype(bool)

    return g_meta, p_meta


# =========================
# HELPERS: CONSISTENCY CHECKS
# =========================
def _compare_meta_rows(a, b, tol=1e-12):
    """
    Compare two dict-like meta rows.
    We require equal H_id and close H01/H02/H12, same seed_H if present, same intransitive if present.
    """
    keys_float = ["H01", "H02", "H12"]
    keys_int = ["H_id", "seed_H"]
    keys_bool = ["intransitive"]

    for k in keys_int:
        if k in a and k in b and pd.notna(a[k]) and pd.notna(b[k]):
            if int(a[k]) != int(b[k]):
                return False

    for k in keys_bool:
        if k in a and k in b and pd.notna(a[k]) and pd.notna(b[k]):
            if bool(a[k]) != bool(b[k]):
                return False

    for k in keys_float:
        if k in a and k in b and pd.notna(a[k]) and pd.notna(b[k]):
            if not np.isfinite(float(a[k])) or not np.isfinite(float(b[k])):
                return False
            if abs(float(a[k]) - float(b[k])) > tol:
                return False

    return True


def check_H_consistency(gauss, plants, g_meta, p_meta, tol=1e-12):
    """
    (i) Check that H entries agree between:
        - gaussian csvs (they include H01/H02/H12)
        - plants csvs (they include H01/H02/H12)
        - gaussian_meta and realdata_meta
    """
    print("\n========== CHECK (i): H consistency ==========")

    # A) gaussian H_id -> representative row
    g_rep = {}
    if not gauss.empty:
        for hid, sub in gauss.groupby("H_id"):
            # take first row
            m = sub.iloc[0].to_dict()
            g_rep[int(hid)] = m

    # B) plants H_id -> representative row (first across all plants)
    p_rep = {}
    if not plants.empty:
        for hid, sub in plants.groupby("H_id"):
            m = sub.iloc[0].to_dict()
            p_rep[int(hid)] = m

    # C) meta maps
    g_meta_map = {}
    if not g_meta.empty and "H_id" in g_meta.columns:
        for hid, sub in g_meta.groupby("H_id"):
            g_meta_map[int(hid)] = sub.iloc[0].to_dict()

    p_meta_map = {}
    if not p_meta.empty and "H_id" in p_meta.columns:
        for hid, sub in p_meta.groupby("H_id"):
            p_meta_map[int(hid)] = sub.iloc[0].to_dict()

    # Union of H ids
    all_h = sorted(set(g_rep.keys()) | set(p_rep.keys()) | set(g_meta_map.keys()) | set(p_meta_map.keys()))
    if not all_h:
        print("[WARN] No H_id found anywhere.")
        return

    mismatches = 0
    missing = 0

    for hid in all_h:
        sources = {}
        if hid in g_rep:
            sources["gaussian_csv"] = g_rep[hid]
        if hid in p_rep:
            sources["plants_csv"] = p_rep[hid]
        if hid in g_meta_map:
            sources["gaussian_meta"] = g_meta_map[hid]
        if hid in p_meta_map:
            sources["plants_meta"] = p_meta_map[hid]

        # if fewer than 2 sources, can't compare
        if len(sources) < 2:
            missing += 1
            print(f"[INFO] H{hid}: only found in {list(sources.keys())}")
            continue

        # compare everything to the first source
        keys = list(sources.keys())
        ref = sources[keys[0]]
        ok_all = True
        for k in keys[1:]:
            if not _compare_meta_rows(ref, sources[k], tol=tol):
                ok_all = False
                break

        if not ok_all:
            mismatches += 1
            print(f"[MISMATCH] H{hid}: sources differ -> {list(sources.keys())}")
            # show values to debug
            for k, row in sources.items():
                show = {kk: row.get(kk, None) for kk in ["H_id", "seed_H", "intransitive", "H01", "H02", "H12"]}
                print(f"  - {k}: {show}")

    print(f"\nSummary: total H={len(all_h)}, mismatches={mismatches}, only-one-source={missing}")
    if mismatches == 0:
        print("[OK] All comparable H entries match within tolerance.")


# =========================
# HELPERS: BINNING + SUMMARIES
# =========================
def summarize_by_bins(df, x_col="std_ratios_C", y_col="alpha_c", p_col="p_c_i",
                      group_cols=("H_id",), n_bins=15):
    """
    Returns a summary with columns:
      group_cols..., x_bin, x_center, y_mean, p_mean
    Log-bins in x_col (strictly >0).
    """
    d = df.copy()

    # Clean x
    d[x_col] = pd.to_numeric(d.get(x_col), errors="coerce")
    d = d.dropna(subset=[x_col])
    d = d[np.isfinite(d[x_col].values)]
    d = d[d[x_col] > 0]
    if d.empty:
        return pd.DataFrame()

    # y and p
    d[y_col] = pd.to_numeric(d.get(y_col), errors="coerce")
    if p_col in d.columns:
        d[p_col] = pd.to_numeric(d[p_col], errors="coerce").round().clip(0, 1)
    else:
        d[p_col] = (~d[y_col].isna()).astype(int)

    # log-bins
    xmin, xmax = float(d[x_col].min()), float(d[x_col].max())
    if (not np.isfinite(xmin)) or (not np.isfinite(xmax)) or xmin <= 0 or xmax <= 0 or xmin == xmax:
        # degenerate -> single bin
        d["x_bin"] = "all"
        d["x_center"] = float(d[x_col].median())
    else:
        edges = np.logspace(np.log10(xmin), np.log10(xmax), n_bins + 1)
        d["x_bin"] = pd.cut(d[x_col], bins=edges, include_lowest=True)
        d = d.dropna(subset=["x_bin"])
        if d.empty:
            return pd.DataFrame()
        intervals = d["x_bin"].cat.categories
        centers = np.sqrt(intervals.left.to_numpy() * intervals.right.to_numpy())
        center_map = {intervals[i]: float(centers[i]) for i in range(len(intervals))}
        d["x_center"] = d["x_bin"].map(center_map)

    group_cols = tuple(group_cols) if group_cols else tuple()
    gb_cols = list(group_cols) + ["x_bin"]

    out = (
        d.groupby(gb_cols, dropna=False)
         .agg(
            x_center=("x_center", "first"),
            y_mean=(y_col, "mean"),
            p_mean=(p_col, "mean"),
         )
         .reset_index()
    )

    sort_cols = list(group_cols) + ["x_center"] if group_cols else ["x_center"]
    return out.sort_values(sort_cols).reset_index(drop=True)


# =========================
# HELPERS: PLOTTING (ROBUST)
# =========================
def plot_summary_perH_generic(summary, y_col, title, ylabel, logx=True, h_col="H_id"):
    if summary is None or summary.empty:
        print(f"[WARN] Empty summary: {title}")
        return
    if y_col not in summary.columns:
        raise KeyError(f"{y_col} not in summary columns: {list(summary.columns)}")

    plt.figure()
    for hid, sub in summary.groupby(h_col):
        sub = sub.sort_values("x_center")
        x = pd.to_numeric(sub["x_center"], errors="coerce").to_numpy(dtype=float)
        y = pd.to_numeric(sub[y_col], errors="coerce").to_numpy(dtype=float)
        ok = np.isfinite(x) & np.isfinite(y)
        if np.any(ok):
            plt.plot(x[ok], y[ok], marker="o", linewidth=1, label=f"H {int(hid)}")

    plt.title(title)
    plt.xlabel("std_ratios_C (bin center)")
    if logx:
        plt.xscale("log")
    plt.ylabel(ylabel)
    plt.grid(True, alpha=0.3)
    plt.legend(ncol=2, fontsize=8)
    plt.tight_layout()
    plt.show()


def plot_summary_pooled_generic(summary, y_col, title, ylabel, logx=True):
    if summary is None or summary.empty:
        print(f"[WARN] Empty summary: {title}")
        return
    if y_col not in summary.columns:
        raise KeyError(f"{y_col} not in summary columns: {list(summary.columns)}")

    plt.figure()
    sub = summary.sort_values("x_center")
    x = pd.to_numeric(sub["x_center"], errors="coerce").to_numpy(dtype=float)
    y = pd.to_numeric(sub[y_col], errors="coerce").to_numpy(dtype=float)

    ok = np.isfinite(x) & np.isfinite(y)
    if np.any(ok):
        plt.plot(x[ok], y[ok], marker="o", linewidth=2)

    plt.title(title)
    plt.xlabel("std_ratios_C (bin center)")
    if logx:
        plt.xscale("log")
    plt.ylabel(ylabel)
    plt.grid(True, alpha=0.3)
    plt.tight_layout()
    plt.show()


# =========================
# MAIN: LOAD DATA
# =========================
os.chdir(BASE_DIR)

gauss = load_gaussian_files(BASE_DIR)
plants = load_plants_files(BASE_DIR, PLANTS)
g_meta, p_meta = load_meta_files(BASE_DIR)

print("Loaded:")
print(f"  gaussian rows: {len(gauss)}")
print(f"  plants rows:   {len(plants)}")
print(f"  g_meta rows:   {len(g_meta)}")
print(f"  p_meta rows:   {len(p_meta)}")

# =========================
# (i) CHECK H CONSISTENCY
# =========================
check_H_consistency(gauss, plants, g_meta, p_meta, tol=H_TOL)


# =========================
# (ii) alpha_c vs std_ratios_C per H separately
# =========================
print("\n========== PLOTS (ii): alpha_c vs std_ratios_C per H ==========")

# GAUSSIAN per H
sum_g_perH = summarize_by_bins(gauss, x_col="std_ratios_C", y_col="alpha_c", p_col="p_c_i",
                               group_cols=("H_id",), n_bins=N_BINS)
plot_summary_perH_generic(sum_g_perH, y_col="y_mean",
                          title="GAUSSIAN: mean alpha_c vs std_ratios_C (per H)",
                          ylabel="mean alpha_c", logx=LOGX)

# PLANTS per H (separate by plant_type)
for plant in PLANTS:
    d = plants[plants["plant_type"] == plant]
    s = summarize_by_bins(d, x_col="std_ratios_C", y_col="alpha_c", p_col="p_c_i",
                          group_cols=("H_id",), n_bins=N_BINS)
    plot_summary_perH_generic(s, y_col="y_mean",
                              title=f"{plant}: mean alpha_c vs std_ratios_C (per H)",
                              ylabel="mean alpha_c", logx=LOGX)


# =========================
# (iii) p_c vs std_ratios_C per H separately
# =========================
print("\n========== PLOTS (iii): p_c vs std_ratios_C per H ==========")

# GAUSSIAN per H
plot_summary_perH_generic(sum_g_perH, y_col="p_mean",
                          title="GAUSSIAN: mean p_c vs std_ratios_C (per H)",
                          ylabel="mean p_c", logx=LOGX)

# PLANTS per H (separate by plant_type)
for plant in PLANTS:
    d = plants[plants["plant_type"] == plant]
    s = summarize_by_bins(d, x_col="std_ratios_C", y_col="alpha_c", p_col="p_c_i",
                          group_cols=("H_id",), n_bins=N_BINS)
    plot_summary_perH_generic(s, y_col="p_mean",
                              title=f"{plant}: mean p_c vs std_ratios_C (per H)",
                              ylabel="mean p_c", logx=LOGX)


# =========================
# (iv) pooled across H (but separated by dataset: gaussian vs each plant type)
# =========================
print("\n========== PLOTS (iv): pooled across H, separated by dataset ==========")

# GAUSSIAN pooled
sum_g_pool = summarize_by_bins(gauss, x_col="std_ratios_C", y_col="alpha_c", p_col="p_c_i",
                               group_cols=(), n_bins=N_BINS)
plot_summary_pooled_generic(sum_g_pool, y_col="y_mean",
                            title="GAUSSIAN (pooled H): mean alpha_c vs std_ratios_C",
                            ylabel="mean alpha_c", logx=LOGX)
plot_summary_pooled_generic(sum_g_pool, y_col="p_mean",
                            title="GAUSSIAN (pooled H): mean p_c vs std_ratios_C",
                            ylabel="mean p_c", logx=LOGX)

# PLANTS pooled (one pooled curve per plant type)
for plant in PLANTS:
    d = plants[plants["plant_type"] == plant]
    s = summarize_by_bins(d, x_col="std_ratios_C", y_col="alpha_c", p_col="p_c_i",
                          group_cols=(), n_bins=N_BINS)
    plot_summary_pooled_generic(s, y_col="y_mean",
                                title=f"{plant} (pooled H): mean alpha_c vs std_ratios_C",
                                ylabel="mean alpha_c", logx=LOGX)
    plot_summary_pooled_generic(s, y_col="p_mean",
                                title=f"{plant} (pooled H): mean p_c vs std_ratios_C",
                                ylabel="mean p_c", logx=LOGX)
```
